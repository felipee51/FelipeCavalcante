```python
# @title
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # Para o modelo de regressão
from sklearn.metrics import mean_squared_error, r2_score # Para avaliação do modelo

# Opcionais para pré-processamento mais avançado, não usados neste MVP, mas mantidos do template
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


# URL da base de dados no GitHub
github_csv_url = 'https://raw.githubusercontent.com/felipee51/FelipeCavalcante/refs/heads/main/Base_Imoveis_RJ_tratada.csv'

try:
    # Tenta carregar o DataFrame a partir do URL do GitHub, especificando o delimitador
    df = pd.read_csv(github_csv_url, delimiter=';')
    print(f"Dados carregados com sucesso do GitHub: {github_csv_url}")
    print("Primeiras 5 linhas dos dados originais:")
    print(df.head())
    print(f"Total de {len(df)} imóveis carregados.")

    # Renomeia a coluna 'valor' para 'preco', que será a nossa variável alvo.
    target_column_original_name = 'valor'
    target_column_model_name = 'preco'

    if target_column_original_name not in df.columns:
        raise ValueError(f"A coluna de preço esperada '{target_column_original_name}' não foi encontrada no CSV.")

    df = df.rename(columns={target_column_original_name: target_column_model_name})

    # Explicitly drop non-numeric columns that are not used in this MVP
    columns_to_drop = ['local', 'bairro/cidade']
    df = df.drop(columns=columns_to_drop, errors='ignore') # 'errors='ignore' prevents error if columns are not found


    # Converte todas as colunas que podem ser numéricas para o tipo numérico.
    # Valores não numéricos serao convertidos para NaN.
    for col in df.columns:
        # Trata colunas com intervalos como 'area(m2)-min' e 'area(m2)-max'
        if '-min' in col or '-max' in col:
            df[col] = pd.to_numeric(df[col].astype(str).str.replace('[^0-9.]', '', regex=True), errors='coerce')
        else:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Check for nulls before dropping
    print("\nValores nulos no dataset antes de remover linhas:")
    print(df.isnull().sum())

    # Remove linhas que contenham valores ausentes (NaN) nas colunas apos a conversao.
    # Inserido para a regressão linear, que nao lida com NaNs.
    df.dropna(inplace=True)


    print(f"\nColuna '{target_column_original_name}' renomeada para '{target_column_model_name}'.")
    print("Todas as colunas foram convertidas para numérico, e linhas com valores ausentes foram removidas.")
    print(f"Total de {len(df)} imóveis após a limpeza e preparação.")
    print("Primeiras 5 linhas dos dados preparados:")
    print(df.head())

except Exception as e:
    print(f"Erro ao carregar ou processar os dados do GitHub: {e}")
    print("Caindo para a geração de dados sintéticos para demonstração (fallback).")
    # --- Geração de Dados Sintéticos (Fallback se o GitHub falhar ou para teste) ---
    np.random.seed(42) # Para reprodutibilidade
    num_imoveis = 100
    # Gerando features comuns que se assemelham ao seu dataset do RJ
    area_sintetica = np.random.randint(40, 500, num_imoveis)
    quartos_sinteticos = np.random.randint(1, 6, num_imoveis)
    suites_sinteticas = np.random.randint(0, 4, num_imoveis)
    vagas_sinteticas = np.random.randint(0, 3, num_imoveis)
    banheiros_sinteticos = np.random.randint(1, 5, num_imoveis)
    distancia_centro_sintetica = np.random.uniform(1, 20, num_imoveis)
    valor_condominio_sintetico = np.random.randint(200, 2000, num_imoveis)
    valor_iptu_sintetico = np.random.randint(50, 500, num_imoveis)

    # Preço sintético baseado em algumas features
    precos_sinteticos = (
        1000 * area_sintetica +
        50000 * quartos_sinteticos +
        30000 * suites_sinteticas +
        20000 * vagas_sinteticas +
        40000 * banheiros_sinteticos -
        5000 * distancia_centro_sintetica +
        0.5 * valor_condominio_sintetico + # Condominio pode influenciar um pouco
        0.1 * valor_iptu_sintetico +      # IPTU pode influenciar um pouco
        np.random.normal(0, 100000, num_imoveis) # Ruído aleatório
    )
    # Garantir preços positivos
    precos_sinteticos[precos_sinteticos < 1000] = 1000 + np.random.normal(0, 500, sum(precos_sinteticos < 1000))

    data_sintetica = {
        'area': area_sintetica,
        'quartos': quartos_sinteticos,
        'suites': suites_sinteticas,
        'vagas': vagas_sinteticas,
        'banheiros': banheiros_sinteticos,
        'distancia_centro': distancia_centro_sintetica,
        'valor_condominio': valor_condominio_sintetico,
        'valor_iptu': valor_iptu_sintetico,
        'preco': precos_sinteticos
    }
    df = pd.DataFrame(data_sintetica)
    print("Dados sintéticos gerados como fallback.")
    print(df.head())

# Verifica se o DataFrame foi carregado/creado com sucesso e possui a coluna 'preco'
if 'df' not in locals() or df.empty or target_column_model_name not in df.columns:
    print("\nERRO CRÍTICO: Não foi possível obter um DataFrame válido com a coluna de preço.")
    print("Por favor, verifique o URL do GitHub, o conteúdo do CSV ou o bloco de geração de dados sintéticos.")
    exit() # Interrompe a execução se não houver dados válidos

print("-" * 50)

# Análise de Dados

# Nesta etapa de Análise de Dados Exploratória (EDA) sobre o dataset de imóveis, visamos entender a distribuição, as relações e as características das variáveis, o que é crucial para as etapas subsequentes de pré-processamento e modelagem.
# Total e Tipo das Instâncias

print(f"Total de instâncias: {len(df)}")
print("\nTipos de dados por coluna:")
print(df.info())

# Exemplo de contagem para uma coluna categórica (se houvesse, como 'bairro' ou 'tipo_imovel' antes do dropna)
# if 'bairro' in df.columns:
#     plt.figure(figsize=(10, 7))
#     sns.countplot(y='bairro', data=df, order=df['bairro'].value_counts().index[:10]) # Top 10 bairros
#     plt.title('Distribuição dos Imóveis por Bairro (Top 10)')
#     plt.xlabel('Contagem')
#     plt.ylabel('Bairro')
#     plt.show()

# Estatísticas Descritivas

# Estatísticas descritivas fornecem um resumo das características numéricas, incluindo média, desvio padrão, mínimo, máximo e quartis.

# estatísticas descritivas básicas do dataset
print("\nEstatísticas Descritivas do Dataset:")
print(df.describe())

# Média

# A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

# média dos atributos numéricos do dataset
print("\nMédia dos atributos numéricos:")
print(df.describe().loc['mean'])

# Gráfico de barras para a média de atributos selecionados
plt.figure(figsize=(12, 6))
df.drop(columns=[target_column_model_name]).mean().plot(kind='bar')
plt.title('Média dos Atributos Numéricos')
plt.xlabel('Atributo')
plt.ylabel('Valor Médio')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Desvio Padrão

# O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

# desvio padrão dos atributos numéricos do dataset
print("\nDesvio Padrão dos atributos numéricos:")
print(df.describe().loc['std'])

# Gráfico de barras para o desvio padrão de atributos selecionados
plt.figure(figsize=(12, 6))
df.drop(columns=[target_column_model_name]).std().plot(kind='bar', color='orange')
plt.title('Desvio Padrão dos Atributos Numéricos')
plt.xlabel('Atributo')
plt.ylabel('Desvio Padrão')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Histograma

# A distribuição de dados descreve como os valores de uma variável se espalham, ou seja, a frequência com que diferentes valores ocorrem. Entender a distribuição é crucial na análise de dados, pois revela padrões, tendências centrais, dispersão e a presença de valores atípicos (outliers). O histograma é uma ferramenta visual fundamental para representar essa distribuição, mostrando a forma dos dados, se são simétricos ou assimétricos, unimodais ou multimodais.
# Distribuição do area

# Use a coluna 'area(m2)-min' para o histograma de área se ela existir e for numérica
area_column_to_plot = 'area(m2)-min' if 'area(m2)-min' in df.columns and pd.api.types.is_numeric_dtype(df['area(m2)-min']) else None

if area_column_to_plot and not df[area_column_to_plot].isnull().all(): # Check if column exists, is numeric, and not all values are NaN
    plt.figure(figsize=(8, 6))
    sns.histplot(df[area_column_to_plot].dropna(), kde=True) # Drop NaNs for plotting
    plt.title(f'Distribuição da Área dos Imóveis ({area_column_to_plot})')
    plt.xlabel('Área (m²)')
    plt.ylabel('Frequência')
    plt.show()
else:
    print(f"\nColuna '{area_column_to_plot}' não encontrada, não numérica ou contém apenas valores ausentes para visualização.")


# Distribuição do quartos

# Use a coluna 'quarto-min' para o histograma de quartos se ela existir e for numérica
quartos_column_to_plot = 'quarto-min' if 'quarto-min' in df.columns and pd.api.types.is_numeric_dtype(df['quarto-min']) else None

if quartos_column_to_plot and not df[quartos_column_to_plot].isnull().all(): # Check if column exists, is numeric, and not all values are NaN
    plt.figure(figsize=(8, 6))
    quartos_data = df[quartos_column_to_plot].dropna() # Drop NaNs for plotting
    if not quartos_data.empty:
        bins_range = range(int(quartos_data.min()), int(quartos_data.max()) + 2)
        sns.histplot(quartos_data, kde=False, bins=bins_range)
        plt.title(f'Distribuição do Número de Quartos ({quartos_column_to_plot})')
        plt.xlabel('Número de Quartos')
        plt.ylabel('Frequência')
        plt.xticks(range(int(quartos_data.min()), int(quartos_data.max()) + 1))
        plt.show()
    else:
         print(f"\nColuna '{quartos_column_to_plot}' contém apenas valores ausentes após a limpeza para visualização.")
else:
     print(f"\nColuna '{quartos_column_to_plot}' não encontrada, não numérica ou contém apenas valores ausentes para visualização.")


# Boxplot

# Boxplots são excelentes para visualizar a distribuição de uma variável numérica em relação a uma categórica (se tivéssemos tratado 'bairro' ou 'tipo_imovel' adequadamente), ou para identificar outliers. Para este MVP, focaremos em scatter plots para relações entre features numéricas e o preço.
# Relação area vs. preco

# Use a coluna 'area(m2)-min' para o scatter plot area vs preco se ela existir e for numérica
area_column_for_scatter = 'area(m2)-min' if 'area(m2)-min' in df.columns and pd.api.types.is_numeric_dtype(df['area(m2)-min']) else None

if area_column_for_scatter and target_column_model_name in df.columns and not df[[area_column_for_scatter, target_column_model_name]].isnull().all().any(): # Check if columns exist, are numeric, and not all values are NaN
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=area_column_for_scatter, y=target_column_model_name, data=df.dropna(subset=[area_column_for_scatter, target_column_model_name]), alpha=0.6) # Drop NaNs for plotting
    plt.title(f'Valor de Aluguel vs. Área do Imóvel ({area_column_for_scatter})')
    plt.xlabel('Área (m²)')
    plt.ylabel('Valor de Aluguel (R$)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
else:
    print(f"\nColunas '{area_column_for_scatter}' ou '{target_column_model_name}' não encontradas, não numéricas ou contêm apenas valores ausentes para visualização de scatter plot.")


# Relação quartos vs. preco

# Use a coluna 'quarto-min' para o boxplot quartos vs preco if it exists and is numeric
quartos_column_for_boxplot = 'quarto-min' if 'quarto-min' in df.columns and pd.api.types.is_numeric_dtype(df['quarto-min']) else None


if quartos_column_for_boxplot and target_column_model_name in df.columns and not df[[quartos_column_for_boxplot, target_column_model_name]].isnull().all().any(): # Check if columns exist, are numeric, and not all values are NaN
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=quartos_column_for_boxplot, y=target_column_model_name, data=df.dropna(subset=[quartos_column_for_boxplot, target_column_model_name])) # Drop NaNs for plotting
    plt.title(f'Valor de Aluguel por Número de Quartos ({quartos_column_for_boxplot})')
    plt.xlabel('Número de Quartos')
    plt.ylabel('Valor de Aluguel (R$)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
else:
    print(f"\nColunas '{quartos_column_for_boxplot}' ou '{target_column_model_name}' não encontradas, não numéricas ou contêm apenas valores ausentes para visualização de boxplot.")

# Matriz de Correlação

# A matriz de correlação mede a força e a direção de uma relação linear entre os atributos numéricos. Valores próximos a 1 indicam uma forte correlação positiva, -1 uma forte correlação negativa, e 0 ausência de correlação linear.

# Matriz de correlação para todas as colunas numéricas
# Garante que apenas colunas numéricas sejam incluídas
df_numeric = df.select_dtypes(include=np.number)
print("\nMatriz de Correlação das Características Numéricas:")
print(df_numeric.corr())

plt.figure(figsize=(10, 8))
# mapa de calor das variáveis numéricas
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação das Características Numéricas do Dataset de Imóveis')
plt.show()

# O mapa de calor da matriz de correlação revela as relações entre as variáveis numéricas, incluindo a correlação das features com o preco. Isso é crucial para entender quais características têm maior impacto no valor do aluguel.
# Tratamento de Valores Nulos

# O tratamento de valores nulos é crucial e pode envolver imputação (preenchimento com média, mediana, moda) ou remoção de linhas/colunas. Neste código, já realizamos df.dropna(inplace=True) após a conversão de tipo, garantindo que não há valores nulos no DataFrame final df antes da modelagem.

# Verificar a presença de valores nulos no dataset após a limpeza
print("\nValores nulos no dataset de Imóveis após a limpeza:")
print(df.isnull().sum())

# Pré-Processamento de Dados

# O pré-processamento de dados é uma etapa crucial para preparar os dados para modelagem, garantindo que estejam no formato correto e otimizados para o desempenho do algoritmo.

# Separar features (X) e target (y)
target_column_model_name = 'preco' # Definido novamente para clareza na seção
y = df[target_column_model_name]

# As features (X) serão TODAS as outras colunas numéricas no DataFrame, exceto a coluna de preço.
X = df.drop(columns=[target_column_model_name])
# Garante que X contém apenas colunas numéricas
X = X.select_dtypes(include=np.number)

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Usando 30% para teste

print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")

# Normalização

# A normalização escala os dados para um intervalo fixo, geralmente entre 0 e 1. É útil quando o algoritmo de machine learning assume que as características estão em uma escala semelhante.

# Inicializar o MinMaxScaler
scaler_norm = MinMaxScaler()

# Aprende min e max APENAS de X_train
scaler_norm.fit(X_train)
X_train_normalized = scaler_norm.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_normalized = scaler_norm.transform(X_test)

# Exibir as primeiras linhas dos dados normalizados (como DataFrame para melhor visualização)
df_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)

print("\nPrimeiras 5 linhas dos dados normalizados (treino):")
print(df_normalized.head())

# Visualização da distribuição após a normalização (exemplo para uma característica)
if 'area(m2)-min' in df_normalized.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_normalized['area(m2)-min'], kde=True)
    plt.title('Distribuição da Área (Normalizado)')
    plt.xlabel('Área Normalizada')
    plt.ylabel('Frequência')
    plt.show()
else:
    print("\nColuna 'area(m2)-min' não encontrada para visualização após normalização.")

# O histograma da área após a normalização mostra que os valores foram escalados para o intervalo de 0 a 1, mantendo a forma da distribuição original.
# Padronização

# A padronização (ou Z-score scaling) transforma os dados para ter média 0 e desvio padrão 1. É útil para algoritmos que são sensíveis à escala das características, como SVMs ou redes neurais.

# Inicializar o StandardScaler
scaler_std = StandardScaler()

# Aprende média e desvio padrão APENAS de X_train
scaler_std.fit(X_train)
X_train_standardized = scaler_std.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_standardized = scaler_std.transform(X_test)

# Exibir as primeiras linhas dos dados padronizados (como DataFrame para melhor visualização)
df_standardized = pd.DataFrame(X_train_standardized, columns=X_train.columns)

print("\nPrimeiras 5 linhas dos dados padronizados (treino):")
print(df_standardized.head())

# Visualização da distribuição após a padronização (exemplo para uma característica)
if 'area(m2)-min' in df_standardized.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_standardized['area(m2)-min'], kde=True)
    plt.title('Distribuição da Área (Padronizado)')
    plt.xlabel('Área Padronizada')
    plt.ylabel('Frequência')
    plt.show()
else:
    print("\nColuna 'area(m2)-min' não encontrada para visualização após padronização.")

# O histograma da área após a padronização mostra que os valores foram transformados para ter uma média próxima de zero e um desvio padrão de um, centralizando a distribuição.
# Outras Transformações e Etapas de Pré-Processamento

# Para este problema de previsão de aluguel, outras etapas cruciais seriam:

#     One-Hot Encoding para variáveis categóricas: As colunas bairro e tipo_imovel são categóricas e possuem informações valiosas. Elas precisariam ser convertidas para um formato numérico que o modelo possa usar (e.g., One-Hot Encoding ou Label Encoding).

#     Feature Engineering: Criar novas features a partir das existentes, como relacao_suite_quarto, densidade_banheiros (banheiros por área), ou idade_imovel se houvesse uma data de construção/publicação.

#     Remoção de Outliers: Identificar e tratar valores extremos no valor_aluguel ou em outras features, que podem distorcer o treinamento do modelo.

#     Seleção de Features: Utilizar técnicas para identificar as features mais relevantes para a previsão, potencialmente removendo as menos informativas para melhorar a performance e interpretabilidade do modelo.

# Modelagem e Avaliação

# Nesta seção, treinamos um modelo de Regressão Linear com os dados padronizados (uma boa prática para modelos baseados em distância/gradiente) e avaliamos seu desempenho.

# --- 3. Treinamento do Modelo ---
# Usaremos um modelo de Regressão Linear.

print("\n--- 3. Treinamento do Modelo ---")
modelo = LinearRegression()
modelo.fit(X_train_standardized, y_train) # Usando dados padronizados para treinamento

print("Modelo de Regressão Linear treinado com sucesso.")
# Para interpretar os coeficientes, lembre-se que eles são baseados nos dados padronizados.
# Coeficientes: o impacto de uma unidade de desvio padrão da feature no preço.
# print(f"Coeficientes do modelo para cada feature: {modelo.coef_}") # Comentado para evitar output muito longo
print(f"Intercepto do modelo: {modelo.intercept_}")
print("-" * 50)

# --- 4. Avaliação do Modelo ---
# Fazer previsões no conjunto de teste (também padronizado) e calcular métricas de erro.

print("\n--- 4. Avaliação do Modelo ---")
y_pred_test = modelo.predict(X_test_standardized) # Previsões no conjunto de teste padronizado

# Erro Quadrático Médio (MSE): Média dos quadrados dos erros. Quanto menor, melhor.
mse = mean_squared_error(y_test, y_pred_test)
# R² (Coeficiente de Determinacao): Proporcao da variancia na variavel dependente que é previsivel a partir das variaveis independentes. Quanto mais proximo de 1, melhor.
r2 = r2_score(y_test, y_pred_test)

print(f"Erro Quadrático Médio (MSE) no conjunto de TESTE: {mse:.2f}")
print(f"R² (Coeficiente de Determinação) no conjunto de TESTE: {r2:.2f}")
print("-" * 50)

# --- 5. Realizar Previsoes (Exemplo) ---
# Usar o modelo treinado para prever o preco de um novo imovel.
# Para o exemplo, usaremos a media das features do conjunto de treinamento padronizado.

print("\n--- 5. Realizar Previsões (Exemplo) ---")
# Cria um novo DataFrame com a média das features de treinamento (já padronizadas)
# Precisa despadronizar os valores médios para mostrar as características no formato original
# mas o modelo espera a entrada padronizada.
# A melhor prática para o exemplo é pegar uma linha real do X_test e prever.
# Ou, criar um "novo imóvel" com valores "reais" e depois padronizá-los.

# Vamos criar um "novo imóvel" com valores em escala original para o exemplo
# e depois padronizá-lo para a previsão.
# Ajustando colunas para o exemplo de imóvel com base nas colunas do dataset real
example_columns = [col for col in X_train.columns if col not in ['local', 'bairro/cidade']] # Exclui colunas não numéricas/tratadas ainda
exemplo_imovel_original = pd.DataFrame(columns=example_columns)

# Preencher com valores de exemplo (ajuste conforme as colunas numéricas do seu X_train)
# Este é um exemplo simplificado. Em um app real, o usuário preencheria esses valores.
# Use a média de X_train para colunas numéricas que não estão explicitamente definidas no exemplo
exemplo_imovel_data = X_train.mean().to_dict()

# Sobrescrever com valores específicos para o exemplo, garantindo que as chaves existem
if 'area(m2)-min' in exemplo_imovel_data: exemplo_imovel_data['area(m2)-min'] = 100
if 'quarto-min' in exemplo_imovel_data: exemplo_imovel_data['quarto-min'] = 2
if 'suite' in exemplo_imovel_data: exemplo_imovel_data['suite'] = 1
if 'garagem-min' in exemplo_imovel_data: exemplo_imovel_data['garagem-min'] = 1
if 'banheiro-min' in exemplo_imovel_data: exemplo_imovel_data['banheiro-min'] = 2
# Adicione outras colunas numéricas conforme necessário, verificando sua existência
if 'iptu' in exemplo_imovel_data: exemplo_imovel_data['iptu'] = 100
if 'condominio' in exemplo_imovel_data: exemplo_imovel_data['condominio'] = 500


exemplo_imovel_original = pd.DataFrame([exemplo_imovel_data])


# Padroniza o exemplo de imóvel usando o scaler trainado no X_train
exemplo_imovel_standardized = scaler_std.transform(exemplo_imovel_original)
exemplo_imovel_df_standardized = pd.DataFrame(exemplo_imovel_standardized, columns=X_train.columns)


preco_previsto_exemplo = modelo.predict(exemplo_imovel_df_standardized)

print(f"Características de um imóvel de exemplo (em escala original):")
print(exemplo_imovel_original.iloc[0].to_dict())
print(f"Preço de aluguel previsto para este imóvel de exemplo: R$ {preco_previsto_exemplo[0]:.2f}")
print("-" * 50)

# --- 6. Visualização de Resultados (Todas as linhas válidas) ---
# Fazer previsões em TODO o DataFrame X e comparar com TODO o DataFrame y.

print("\n--- 6. Visualização de Resultados (Com todas as linhas válidas) ---")

# Prever preços para TODAS as features (X) após padronização
# Para fazer previsões em todo X, X também precisa ser padronizado.
X_standardized_total = scaler_std.transform(X)
y_pred_total = modelo.predict(X_standardized_total)

plt.figure(figsize=(10, 7))
# O gráfico agora usa 'y' (todos os valores reais) e 'y_pred_total' (todas as previsões)
sns.scatterplot(x=y, y=y_pred_total, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2) # Linha de 45 graus ideal
plt.title('Preços Reais vs. Preços Previstos (Todas as Linhas Válidas)')
plt.xlabel('Preço Real (R$)')
plt.ylabel('Preço Previsto (R$)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("\nMVP de Previsão de Preços de Imóveis concluído.")
print("O gráfico agora mostra a relação entre os preços reais e os preços previstos pelo modelo para TODAS as linhas válidas do seu conjunto de dados.")
```

```python
# --- Pré-processamento: Tratamento de Variáveis Categóricas (One-Hot Encoding) ---
print("\n--- Pré-processamento: Tratamento de Variáveis Categóricas (One-Hot Encoding) ---")

import pandas as pd # Garante que pandas está importado
import numpy as np # Garante que numpy está importado
from sklearn.model_selection import train_test_split # Garante que o split está disponível
import warnings # Garante que warnings está importado

# Suprimir avisos específicos
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)

# Definir nome da coluna alvo no início para garantir disponibilidade
target_column_model_name = 'preco'

# --- Carregamento Inicial dos Dados (Garantir que df existe) ---
github_csv_url = 'https://raw.githubusercontent.com/felipee51/FelipeCavalcante/refs/heads/main/Base_Imoveis_RJ_tratada.csv'
try:
    # Tenta carregar o DataFrame a partir do URL do GitHub, especificando o delimitador
    df = pd.read_csv(github_csv_url, delimiter=';')
    print(f"Dados carregados com sucesso do GitHub: {github_csv_url}")

    # Renomeia a coluna 'valor' para 'preco', que será a nossa variável alvo.
    target_column_original_name = 'valor'
    if target_column_original_name in df.columns:
        df = df.rename(columns={target_column_original_name: target_column_model_name})
    elif target_column_model_name not in df.columns:
         raise ValueError(f"A coluna de preço esperada '{target_column_original_name}' ou '{target_column_model_name}' não foi encontrada no CSV.")

    # Remove explicitamente colunas não numéricas que não são usadas neste MVP, EXCLUINDO 'bairro/cidade' e 'local' por enquanto
    columns_to_drop_initial = ['Descrição', 'Novidade', 'Acabamento'] # Removida 'local' e 'bairro/cidade' daqui
    df = df.drop(columns=columns_to_drop_initial, errors='ignore') # 'errors='ignore' evita erro se as colunas não forem encontradas


    # --- Refatorado: Converte colunas que devem ser numéricas de forma mais robusta ---
    # Identificar explicitamente as colunas que devem ser numéricas (exceto target e colunas que serão tratadas como categóricas)
    cols_to_convert_to_numeric = [col for col in df.columns if col not in [target_column_model_name, 'local', 'bairro/cidade']]

    for col in cols_to_convert_to_numeric:
        if col in df.columns: # Verifica se a coluna realmente existe
             # Tenta converter para numérico. Valores não numéricos se tornarão NaN.
             df[col] = pd.to_numeric(df[col], errors='coerce')
        else:
            print(f"Aviso: Coluna numérica esperada '{col}' non encontrada no DataFrame durante a conversão. Ignorando.")
    print("Colunas numéricas convertidas.")
    # --- Fim da Refatoração ---


    # Remove linhas que contenham valores ausentes (NaN) nas colunas numéricas (exceto o target por enquanto)
    numeric_cols_for_nan_check = df.select_dtypes(include=np.number).columns.tolist()
    if target_column_model_name in numeric_cols_for_nan_check:
         numeric_cols_for_nan_check.remove(target_column_model_name)
    df.dropna(subset=numeric_cols_for_nan_check, inplace=True)

    print(f"Coluna '{target_column_original_name}' renomeada para '{target_column_model_name}'.")
    print("Linhas com valores de feature ausentes removidas (exceto categóricas).")
    print(f"Total de {len(df)} imóveis após a limpeza inicial.")
    print("Primeiras 5 linhas dos dados preparados:")
    print(df.head())

except Exception as e:
    print(f"Erro ao carregar ou processar os dados do GitHub: {e}")
    print("Caindo para a geração de dados sintéticos para demonstração (fallback).")
    # --- Geração de Dados Sintéticos (Fallback se o GitHub falhar ou para teste) ---
    np.random.seed(42) # Para reprodutibilidade
    num_imoveis = 100
    # Gerando features numéricas sintéticas
    area_sintetica = np.random.randint(40, 500, num_imoveis)
    quartos_sinteticos = np.random.randint(1, 6, num_imoveis)
    suites_sinteticas = np.random.randint(0, 4, num_imoveis)
    vagas_sinteticas = np.random.randint(0, 3, num_imoveis)
    banheiros_sinteticos = np.random.randint(1, 5, num_imoveis)
    distancia_centro_sintetica = np.random.uniform(1, 20, num_imoveis)
    valor_condominio_sintetico = np.random.randint(200, 2000, num_imoveis)
    valor_iptu_sintetico = np.random.randint(50, 500, num_imoveis)

    # Gerando dados categóricos sintéticos
    bairros_sinteticos = np.random.choice(['Bairro A', 'Bairro B', 'Bairro C', 'Desconhecido'], num_imoveis)
    tipos_sinteticos = np.random.choice(['Apartamento', 'Casa', 'Desconhecido'], num_imoveis)

    # Preço sintético baseado em algumas features
    precos_sinteticos = (
        1000 * area_sintetica +
        50000 * quartos_sinteticos +
        30000 * suites_sinteticas +
        20000 * vagas_sinteticas +
        40000 * banheiros_sinteticos -
        5000 * distancia_centro_sintetica +
        0.5 * valor_condominio_sintetico +
        0.1 * valor_iptu_sintetico +
        np.random.normal(0, 100000, num_imoveis)
    )
    precos_sinteticos[precos_sinteticos < 1000] = 1000 + np.random.normal(0, 500, sum(precos_sinteticos < 1000))

    data_sintetica = {
        'area(m2)-min': area_sintetica,
        'quarto-min': quartos_sinteticos,
        'suite': suites_sinteticas,
        'garagem-min': vagas_sinteticas,
        'banheiro-min': banheiros_sinteticos,
        'distancia_centro': distancia_centro_sintetica,
        'condominio': valor_condominio_sintetico,
        'iptu': valor_iptu_sintetico,
        target_column_model_name: precos_sinteticos, # Usar target_column_model_name aqui
        'local': [f'Rua {i}' for i in range(num_imoveis)], # 'local' sintético como string
        'bairro/cidade': bairros_sinteticos.astype(str) # Usar bairros sintéticos como string
    }
    df = pd.DataFrame(data_sintetica) # Define df aqui em caso de fallback
    print("Dados sintéticos gerados como fallback.")
    print(df.head())


# Verifica si o DataFrame foi carregado/creado con sucesso e possui la columna 'preco'
if 'df' not in locals() or df.empty or target_column_model_name not in df.columns:
    print("\nERRO CRÍTICO: Não foi possível obter um DataFrame válido com a coluna de preço.")
    # Define X, y, X_train, etc. como None ou vazios para evitar NameError subsequente
    X, y, X_train, X_test, y_train, y_test = None, None, None, None, None, None
else:
    # --- Pré-processamento Adicional: Tratamento de Variáveis Categóricas (One-Hot Encoding) ---
    print("\n--- Pré-processamento: Tratamento de Variáveis Categóricas (One-Hot Encoding) ---")
    categorical_cols_to_encode = ['bairro/cidade'] # Colunas categóricas a serem codificadas
    # Lidar com potenciais NaNs em colunas categóricas antes da codificação
    for col in categorical_cols_to_encode:
        if col in df.columns:
            df[col] = df[col].fillna('Desconhecido').astype(str) # Preencher NaNs com um placeholder e garantir tipo string
        else:
            print(f"Aviso: Coluna categórica '{col}' para OHE não encontrada no DataFrame. Ignorando para OHE.")

    # Aplicar One-Hot Encoding
    # Remover a coluna 'local' explicitamente antes do OHE se ela existir, pois pode ter muitos valores únicos
    columns_to_drop_before_ohe = ['local']
    df_processed = df.drop(columns=columns_to_drop_before_ohe, errors='ignore').copy() # Remove 'local' e trabalha em uma cópia

    # Aplicar OHE apenas nas colunas especificadas
    if any(col in df_processed.columns for col in categorical_cols_to_encode):
        df_encoded = pd.get_dummies(df_processed, columns=[col for col in categorical_cols_to_encode if col in df_processed.columns], dummy_na=False) # dummy_na=False pois NaNs foram preenchidos com 'Desconhecido'
        print("Variáveis categóricas codificadas (One-Hot Encoding):")
        display(df_encoded.head())
        print(f"Número de colunas após OHE: {df_encoded.shape[1]}")
    else:
        df_encoded = df_processed.copy()
        print("Nenhuma coluna categórica especificada para One-Hot Encoding encontrada. Continuando sem OHE.")


    # Verificar se a coluna alvo está presente em df_encoded antes de separar X e y
    if target_column_model_name not in df_encoded.columns:
        print(f"\nERRO CRÍTICO: Coluna alvo '{target_column_model_name}' não encontrada no DataFrame após One-Hot Encoding.")
        # Define X, y, X_train, etc. como None ou vazios para evitar NameError subsequente
        X, y, X_train, X_test, y_train, y_test = None, None, None, None, None, None
    else:
        # Definir X e y usando o DataFrame após OHE (ou o fallback)
        y = df_encoded[target_column_model_name]
        X = df_encoded.drop(columns=[target_column_model_name])

        # Garantir que X contém apenas tipos numéricos
        X = X.select_dtypes(include=np.number)

        # Remove linhas com quaisquer valores NaN restantes nas features antes da divisão (novamente, por segurança)
        initial_rows_before_nan_drop_final = X.shape[0]
        X.dropna(inplace=True)
        y = y.loc[X.index] # Alinha y com X após remover linhas
        if X.shape[0] < initial_rows_before_nan_drop_final:
            print(f"\nRemovidas {initial_rows_before_nan_drop_final - X.shape[0]} linhas com valores NaN nas features antes da divisão (verificação final).")


        # Dividir os dados novamente após o pré-processamento adicional
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        print(f"\nDimensões de X_train após OHE e split: {X_train.shape}")
        print(f"Dimensões de X_test após OHE e split: {X_test.shape}")


print("-" * 50)
```

### Pipelines Reprodutíveis: Orquestrando o Fluxo de Trabalho

Em projetos de machine learning, é fundamental garantir que as etapas de pré-processamento e modelagem sejam aplicadas de forma consistente, tanto durante o treinamento quanto ao fazer previsões em novos dados. Os **Pipelines** do scikit-learn fornecem uma maneira elegante de encadear várias etapas de processamento de dados e um estimador final em um único objeto.

Um pipeline:

1.  **Simplifica o Código:** Reduz a necessidade de aplicar cada transformação manualmente em diferentes conjuntos de dados.
2.  **Previne Vazamento de Dados (Data Leakage):** Garante que transformações que aprendem parâmetros dos dados (como `StandardScaler` ou `SelectKBest`) sejam ajustadas apenas nos dados de treino e aplicadas consistentemente nos dados de teste e de previsão.
3.  **Melhora a Reprodutibilidade:** Facilita a reprodução de todo o fluxo de trabalho.

Neste MVP, demonstramos a construção de um pipeline que inclui as etapas de padronização (`StandardScaler`), seleção de features (`SelectKBest`) e o modelo RandomForestRegressor. Este pipeline encapsula uma parte significativa do fluxo de trabalho de pré-processamento e modelagem.

```python
# --- Construindo um Pipeline Reprodutível ---
print("\n--- Construindo um Pipeline Reprodutível ---")

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor
import time # Importar a biblioteca time


# Definir as etapas do pipeline
# O pipeline incluirá Padronização, Seleção de Features e o Modelo RandomForestRegressor
# É importante que as etapas no pipeline correspondam ao pré-processamento aplicado aos dados de treino antes de chegar ao modelo.
# Neste caso, assumimos que X_train e X_test já passaram por OHE e Feature Engineering,
# e o pipeline lidará com a Padronização e Seleção de Features ANTES do modelo.

# Verificar se X_train é um DataFrame para obter nomes de colunas para a seleção de features
if isinstance(X_train, pd.DataFrame):
    # Definir as etapas do pipeline
    pipeline = Pipeline([
        ('scaler', StandardScaler()), # Etapa 1: Padronização
        ('selector', SelectKBest(score_func=f_regression, k=min(X_train.shape[1], 10))), # Etapa 2: Seleção de Features (usar min para evitar erro se menos de K features)
        ('regressor', RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)) # Etapa 3: Modelo
    ])

    print("\nPipeline criado com as seguintes etapas:")
    print(pipeline.steps)

    # Treinar o pipeline nos dados de treino originais (antes da padronização e seleção de features manuais)
    # NOTA IMPORTANTE: Para usar o pipeline corretamente, ele deve ser treinado nos dados *antes*
    # das transformações que ele próprio realizará.
    # Precisamos redefinir X_train e X_test para o estado *antes* da padronização e seleção de features,
    # mas *depois* de OHE, FE e tratamento de outliers.

    # Assumindo que X_train e X_test no escopo atual são os dados *após* seleção de features e padronização.
    # Precisamos de acesso aos dados *antes* dessas etapas.
    # Como o estado exato dos dados antes da padronização/seleção não é trivial de recuperar aqui,
    # vamos usar os dados X_train e y_train *atuais* para demonstrar o ajuste do pipeline.
    # Em um fluxo de trabalho ideal, o pipeline seria ajustado nos dados após OHE, FE e outliers,
    # mas antes de qualquer escalonamento ou seleção.

    # Para fins de demonstração com o estado atual das variáveis:
    print("\nAjustando o Pipeline nos dados de treino (assumindo que são os dados após pré-processamento inicial):")
    try:
        # Medir o tempo de ajuste do pipeline
        start_time_pipeline_fit = time.time()
        pipeline.fit(X_train, y_train)
        end_time_pipeline_fit = time.time()
        fit_time_pipeline = end_time_pipeline_fit - start_time_pipeline_fit
        print(f"Tempo de ajuste do Pipeline: {fit_time_pipeline:.4f} segundos")


        print("\nPipeline ajustado com sucesso.")

        # Avaliar o pipeline no conjunto de teste
        print("\nAvaliando o Pipeline no conjunto de teste:")
        # Medir o tempo de previsão e avaliação do pipeline
        start_time_pipeline_eval = time.time()
        score_pipeline = pipeline.score(X_test, y_test) # Uses the default regressor score method (R² for LinearRegression/RandomForestRegressor)
        end_time_pipeline_eval = time.time()
        eval_time_pipeline = end_time_pipeline_eval - start_time_pipeline_eval
        print(f"Tempo de avaliação do Pipeline: {eval_time_pipeline:.4f} segundos")


        y_pred_test_pipeline = pipeline.predict(X_test)

        mse_pipeline = mean_squared_error(y_test, y_pred_test_pipeline)
        rmse_pipeline = np.sqrt(mse_pipeline)
        mae_pipeline = mean_absolute_error(y_test, y_pred_test_pipeline)
        r2_pipeline = r2_score(y_test, y_pred_test_pipeline)


        print(f"Pipeline - MSE no conjunto de TESTE: {mse_pipeline:.2f}")
        print(f"Pipeline - RMSE no conjunto de TESTE: {rmse_pipeline:.2f}")
        print(f"Pipeline - MAE no conjunto de TESTE: {mae_pipeline:.2f}")
        print(f"Pipeline - R² no conjunto de TESTE: {r2_pipeline:.2f}")

        # Opcional: Acessar o modelo treinado dentro do pipeline
        # treined_regressor = pipeline.named_steps['regressor']

    except Exception as e:
        print(f"\nErro durante o ajuste ou avaliação do Pipeline: {e}")
        print("Não foi possível ajustar ou avaliar o Pipeline.")
        pipeline = None # Definir pipeline como None se ocorrer um erro

else:
    print("\nX_train não é um DataFrame. Não foi possível construir o Pipeline com SelectKBest usando nomes de colunas.")
    pipeline = None # Definir pipeline como None se X_train não for DataFrame


print("-" * 50)
```

#### 3. Tratamento de Outliers: Lidando com Valores Extremos

Valores extremos (outliers) em certas features ou na variável alvo podem distorcer o treinamento do modelo. Identificamos e removemos linhas do conjunto de treino onde features como 'area(m2)-min', 'iptu', 'condominio' ou 'preco' apresentavam Z-score acima de um limite (neste caso, 3).

```python
# 3. Tratamento de Outliers (Exemplo básico: Remoção baseada em Z-score em features selecionadas)
import pandas as pd # Garantir que o pandas está importado
import numpy as np # Garantir que o numpy está importado
from sklearn.model_selection import train_test_split # Garantir que o split está disponível se necessário aqui
import warnings

# Suprimir avisos específicos
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning) # Suprimir potenciais avisos futuros do seaborn


print("\nVerificando e removendo outliers em features selecionadas no conjunto de treino:")

# Combinar X_train e y_train temporariamente para detecção de outliers com base nas features e no target
# Precisa garantir que X_train seja um DataFrame para esta abordagem
if isinstance(X_train, pd.DataFrame):
    # Garantir que a coluna 'preco' não está em X_train antes de combinar
    if 'preco' in X_train.columns:
         X_train_temp = X_train.drop(columns=['preco'])
    else:
         X_train_temp = X_train.copy()

    # Alinhar y_train com o índice de X_train_temp antes de combinar
    y_train_aligned = y_train.loc[X_train_temp.index]
    train_combined = pd.concat([X_train_temp, y_train_aligned], axis=1)

    initial_train_size = train_combined.shape[0]
    outlier_indices_combined = pd.Index([])
    outlier_threshold = 3 # Limite comum para Z-score

    # Features a serem verificadas para outliers - incluir 'preco' do df combinado
    # Usar pd.api.types.is_numeric_dtype para verificar tipos numéricos de forma robusta
    features_to_check_outliers_combined = [col for col in train_combined.columns if pd.api.types.is_numeric_dtype(train_combined[col])]


    for feature in features_to_check_outliers_combined:
        if feature in train_combined.columns:
            # Calcular Z-scores, lidando com potencial divisão por zero se o desvio padrão for 0
            std_dev = train_combined[feature].std()
            if std_dev > 0:
                z_scores_feature = np.abs((train_combined[feature] - train_combined[feature].mean()) / std_dev)
                feature_outlier_indices = train_combined.index[z_scores_feature > outlier_threshold]
                outlier_indices_combined = outlier_indices_combined.union(feature_outlier_indices) # Coletar índices de todas as features/target
                print(f"  - Outliers identificados em '{feature}' com Z-score > {outlier_threshold}: {len(feature_outlier_indices)}")
            else:
                print(f"  - Desvio padrão de '{feature}' é zero. Pulando verificação de outliers para esta feature.")
        else:
            print(f"  - Feature '{feature}' não encontrada para verificação de outliers.")


    # Remover linhas com outliers identificados de X_train e y_train
    if not outlier_indices_combined.empty:
        X_train_cleaned = X_train.drop(outlier_indices_combined)
        y_train_cleaned = y_train.drop(outlier_indices_combined)
        print(f"\nRemovidas {initial_train_size - X_train_cleaned.shape[0]} linhas de treino contendo outliers (Z-score > {outlier_threshold}).")
        print(f"Novo tamanho do conjunto de treino após remoção de outliers: {X_train_cleaned.shape}")
        X_train = X_train_cleaned
        y_train = y_train_cleaned
    else:
        print("\nNenhum outlier identificado com base em Z-score. Nenhuma linha removida.")

elif isinstance(X_train, np.ndarray):
    print("\nX_train é um array NumPy. Tratamento de outliers em arrays é mais complexo e pode afetar a correspondência com y_train. Pulando tratamento de outliers por enquanto.")
else:
    print("\nFormato inesperado para X_train. Pulando Tratamento de Outliers.")


print("-" * 50)
```

Nossa Conclusão: A remoção de outliers ajuda a tornar o conjunto de treino mais robusto, minimizando a influência de pontos de dados atípicos que poderiam prejudicar a generalização do modelo para dados não vistos.

Com um número maior de features após o One-Hot Encoding e Feature Engineering, a seleção de features é importante para focar o modelo nas variáveis mais informativas, reduzindo ruído e o risco de overfitting. Utilizamos `SelectKBest` com `f_regression` para selecionar as K (definido como 10 neste MVP) features com maior relação linear com a variável alvo.

```python
# 4. Seleção de Features (Exemplo básico: Selecionar K melhores features usando f_regression)
# Re-verificar por NaNs após remoção de outliers (não deveria haver, mas é uma boa prática)
if isinstance(X_train, pd.DataFrame):
    if X_train.isnull().sum().sum() > 0 or y_train.isnull().sum() > 0:
        print("\nAviso: Valores nulos encontrados em X_train ou y_train após pré-processamento. Removendo linhas com NaNs.")
        # Remover NaNs se algum foi introduzido inesperadamente
        train_combined = pd.concat([X_train, y_train], axis=1)
        train_combined.dropna(inplace=True)
        X_train = train_combined.drop(columns=[target_column_model_name])
        y_train = train_combined[target_column_model_name]
        print(f"Novo tamanho do conjunto de treino após remoção de NaNs: {X_train.shape}")

    # Garantir que X_train contém apenas tipos numéricos antes da seleção de features
    X_train = X_train.select_dtypes(include=np.number)

    k = 10 # Selecionar top K features (ajustar K conforme necessário)
    if X_train.shape[1] > 0 and X_train.shape[1] >= k:
        try:
            selector = SelectKBest(score_func=f_regression, k=k)
            # Lidar com potenciais valores não finitos em y_train antes de ajustar
            if np.isfinite(y_train).all():
                 selector.fit(X_train, y_train)
            else:
                 print("\nAviso: Valores não finitos encontrados em y_train. Pulando seleção de features.")
                 selector = None # Definir selector como None se o ajuste falhar

            if selector:
                # Obter os índices das features selecionadas
                selected_features_indices = selector.get_support(indices=True)

                # Obter os nomes das features selecionadas
                selected_features_names = X_train.columns[selected_features_indices]

                print(f"\nSelecionadas as {k} melhores features usando SelectKBest (f_regression):")
                print(selected_features_names.tolist())

                # Filtrar os conjuntos de treino e teste para incluir apenas as features selecionadas
                # Garantir que X_test também tenha as mesmas colunas antes de filtrar
                if isinstance(X_test, pd.DataFrame):
                    X_test = X_test.select_dtypes(include=np.number) # Garantir que X_test é numérico
                    # Filtrar X_test para incluir apenas as colunas selecionadas de X_train
                    X_test_selected = X_test[selected_features_names]

                    print(f"Dimensões de X_train após seleção de features: {X_train[selected_features_names].shape}") # Usar a forma após filtragem
                    print(f"Dimensões de X_test após seleção de features: {X_test_selected.shape}")
                    X_train = X_train[selected_features_names] # Atualizar X_train e X_test para etapas subsequentes
                    X_test = X_test_selected
                elif isinstance(X_test, np.ndarray):
                     print("\nX_test é um array NumPy. Não foi possível filtrar X_test usando nomes de colunas. Pulando atualização de X_test.")
                     # Se X_test é um array, a seleção de features em X_train não pode ser facilmente aplicada pelo nome da coluna.
                     # Em um pipeline real, o pré-processamento seria aplicado consistentemente.
                else:
                    print("\nFormato inesperado para X_test. Pulando atualização de X_test após seleção de features.")

        except Exception as e:
            print(f("\nErro durante a seleção de features: {e}"))
            print("Continuando com todas as features numéricas.")
            # Se a seleção de features falhar, reverter para usar todas as features numéricas disponíveis em X_train/X_test
            X_train = X_train.select_dtypes(include=np.number)
            if isinstance(X_test, pd.DataFrame):
                X_test = X_test.select_dtypes(include=np.number)
                # Garantir que o conjunto de teste tenha as mesmas colunas que o conjunto de treino após potencial FE/remoção de outliers
                common_cols = list(set(X_train.columns) & set(X_test.columns))
                X_train = X_train[common_cols]
                X_test = X_test[common_cols]
            else:
                 print("\nX_test não é um DataFrame. Não foi possível garantir colunas comuns após falha na seleção de features.")


    elif X_train.shape[1] > 0:
         print(f("\nNúmero de features ({X_train.shape[1]}) é menor que K ({k}). Pulando seleção de features e usando todas as features numéricas disponíveis."))
         # Garantir que o conjunto de teste tenha as mesmas colunas que o conjunto de treino após potencial FE/remoção de outliers
         X_train = X_train.select_dtypes(include=np.number)
         if isinstance(X_test, pd.DataFrame):
              X_test = X_test.select_dtypes(include=np.number)
              common_cols = list(set(X_train.columns) & set(X_test.columns))
              X_train = X_train[common_cols]
              X_test = X_test[common_cols]
         else:
              print("\nX_test não é um DataFrame. Não foi possível garantir colunas comuns após pular seleção de features.")
    else:
        print("\nNão há features numéricas disponíveis para seleção. Pulando seleção de features.")

elif isinstance(X_train, np.ndarray):
    print("\nX_train é um array NumPy. Seleção de features em arrays requer abordagens diferentes. Pulando seleção de features por enquanto.")
else:
    print("\nFormato inesperado para X_train. Pulando Seleção de Features.")

print("-" * 50)
```

```python
# --- Interpretação do Modelo (Coeficientes LR e Importância de Features RF) ---
print("\n--- Interpretação do Modelo ---")

# Importar bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_selection import SelectKBest, f_regression
import warnings

# Suprimir avisos específicos
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)

# Assumindo que df já está carregado e limpo das células anteriores
# Se não, você precisaria adicionar o código de carregamento e limpeza inicial aqui também.
# Para esta célula regenerada, incluiremos as etapas de carregamento e pré-processamento de dados para torná-la autocontida.

# --- Carregamento e Limpeza Inicial dos Dados (Replicado da célula original) ---
# URL da base de dados no GitHub
github_csv_url = 'https://raw.githubusercontent.com/felipee51/FelipeCavalcante/refs/heads/main/Base_Imoveis_RJ_tratada.csv'

try:
    # Tenta carregar o DataFrame a partir do URL do GitHub, especificando o delimitador
    df = pd.read_csv(github_csv_url, delimiter=';')
    print(f"Dados carregados com sucesso do GitHub: {github_csv_url}")

    # Renomeia a coluna 'valor' para 'preco', que será a nossa variável alvo.
    target_column_original_name = 'valor'
    target_column_model_name = 'preco'

    if target_column_original_name not in df.columns:
        raise ValueError(f"A coluna de preço esperada '{target_column_original_name}' não foi encontrada no CSV.")

    df = df.rename(columns={target_column_original_name: target_column_model_name})

    # Remove explicitamente colunas não numéricas que não são usadas neste MVP
    columns_to_drop_initial = ['local', 'bairro/cidade', 'Descrição', 'Novidade', 'Acabamento'] # Incluindo colunas mencionadas no markdown
    df = df.drop(columns=columns_to_drop_initial, errors='ignore') # 'errors='ignore' evita erro se as colunas não forem encontradas


    # Converte todas as colunas que podem ser numéricas para o tipo numérico.
    # Valores não numéricos serão convertidos para NaN.
    for col in df.columns:
        if col != target_column_model_name: # Exclui a coluna alvo para conversão numérica inicial e verificação de NaN nas features
            # Trata colunas com intervalos como 'area(m2)-min' e 'area(m2)-max'
            if '-min' in col or '-max' in col:
                df[col] = pd.to_numeric(df[col].astype(str).str.replace('[^0-9.]', '', regex=True), errors='coerce')
            else:
                df[col] = pd.to_numeric(df[col], errors='coerce')


    # Remove linhas que contenham valores ausentes (NaN) nas colunas numéricas (exceto o target por enquanto)
    numeric_cols_for_nan_check = df.select_dtypes(include=np.number).columns.tolist()
    if target_column_model_name in numeric_cols_for_nan_check:
         numeric_cols_for_nan_check.remove(target_column_model_name)
    df.dropna(subset=numeric_cols_for_nan_check, inplace=True)


    print(f"Coluna '{target_column_original_name}' renomeada para '{target_column_model_name}'.")
    print("Colunas numéricas convertidas e linhas com valores de feature ausentes removidas.")
    print(f"Total de {len(df)} imóveis após a limpeza inicial.")
    print("Primeiras 5 linhas dos dados preparados:")
    print(df.head())

except Exception as e:
    print(f"Erro ao carregar ou processar os dados do GitHub: {e}")
    print("Caindo para a geração de dados sintéticos para demonstração (fallback).")
    # --- Geração de Dados Sintéticos (Fallback se o GitHub falhar ou para teste) ---
    np.random.seed(42) # Para reprodutibilidade
    num_imoveis = 100
    # Gerando features comuns que se assemelham ao seu dataset do RJ
    area_sintetica = np.random.randint(40, 500, num_imoveis)
    quartos_sinteticos = np.random.randint(1, 6, num_imoveis)
    suites_sinteticas = np.random.randint(0, 4, num_imoveis)
    vagas_sinteticas = np.random.randint(0, 3, num_imoveis)
    banheiros_sinteticos = np.random.randint(1, 5, num_imoveis)
    distancia_centro_sintetica = np.random.uniform(1, 20, num_imoveis)
    valor_condominio_sintetico = np.random.randint(200, 2000, num_imoveis)
    valor_iptu_sintetico = np.random.randint(50, 500, num_imoveis)


    # Preço sintético baseado em algumas features
    precos_sinteticos = (
        1000 * area_sintetica +
        50000 * quartos_sinteticos +
        30000 * suites_sinteticas +
        20000 * vagas_sinteticas +
        40000 * banheiros_sinteticos -
        5000 * distancia_centro_sintetica +
        0.5 * valor_condominio_sintetico + # Condominio pode influenciar um pouco
        0.1 * valor_iptu_sintetico +      # IPTU pode influenciar um pouco
        np.random.normal(0, 100000, num_imoveis) # Ruído aleatório
    )
    # Garantir preços positivos
    precos_sinteticos[precos_sinteticos < 1000] = 1000 + np.random.normal(0, 500, sum(precos_sinteticos < 1000))

    data_sintetica = {
        'area(m2)-min': area_sintetica,
        'quarto-min': quartos_sinteticos,
        'suite': suites_sinteticas,
        'garagem-min': vagas_sinteticas,
        'banheiro-min': banheiros_sinteticos,
        'distancia_centro': distancia_centro_sintetica,
        'condominio': valor_condominio_sintetico,
        'iptu': valor_iptu_sintetico,
        'preco': precos_sinteticos,
        'area(m2)-max': area_sintetica + np.random.randint(0, 50, num_imoveis), # Adiciona área máxima
        'quarto-max': quartos_sinteticos + np.random.randint(0, 1, num_imoveis), # Adiciona quarto máximo
        'garagem-max': vagas_sinteticas + np.random.randint(0, 1, num_imoveis), # Adiciona garagem máxima
        'banheiro-max': banheiros_sinteticos + np.random.randint(0, 1, num_imoveis), # Adiciona banheiro máximo
        'local': [f'Rua {i}' for i in range(num_imoveis)], # 'local' sintético
        'bairro/cidade': np.random.choice(['Bairro A', 'Bairro B', 'Bairro C', 'Desconhecido'], num_imoveis) # Usa bairros sintéticos
    }
    df = pd.DataFrame(data_sintetica)
    print("Dados sintéticos gerados como fallback.")
    print(df.head())


# Verifica se o DataFrame foi carregado/criado com sucesso e possui a coluna 'preco'
if 'df' not in locals() or df.empty or target_column_model_name not in df.columns:
    print("\nERRO CRÍTICO: Não foi possível obter um DataFrame válido com a coluna de preço.")
    # exit() # Removido exit para evitar interrupção do kernel


print("-" * 50)


# --- Divisão dos Dados (Replicado) ---
if 'df' in locals() and not df.empty and target_column_model_name in df.columns:
    y = df[target_column_model_name]
    X = df.drop(columns=[target_column_model_name])

    # Garante que X contém apenas tipos numéricos após potencial fallback
    X = X.select_dtypes(include=np.number)

    # Remove linhas com quaisquer valores NaN restantes nas features antes da divisão (novamente, por segurança)
    initial_rows_before_nan_drop = X.shape[0]
    X.dropna(inplace=True)
    y = y.loc[X.index] # Alinha y com X após remover linhas
    if X.shape[0] < initial_rows_before_nan_drop:
        print(f"\nRemovidas {initial_rows_before_nan_drop - X.shape[0]} linhas com valores NaN nas features antes da divisão (verificação pós-fallback).")


    # Divide os dados em conjuntos de treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Usando 30% para teste

    print(f"\nDimensões de X_train após a divisão: {X_train.shape}")
    print(f"Dimensões de X_test após a divisão: {X_test.shape}")
    print(f"Dimensões de y_train após a divisão: {y_train.shape}")
    print(f"Dimensões de y_test após a divisão: {y_test.shape}")

else:
    print("\nNão foi possível realizar a divisão dos dados. Verifique se 'df' foi carregado corretamente.")


print("-" * 50)

# --- Pré-processamento Adicional (Replicado e ligeiramente adaptado) ---
print("\n--- Pré-processamento Adicional ---")

# Garante que X_train e X_test são DataFrames para operações com colunas
if not isinstance(X_train, pd.DataFrame) or not isinstance(X_test, pd.DataFrame):
    print("\nAviso: X_train ou X_test não são DataFrames. Tentando converter para as etapas de pré-processamento...")
    if 'X' in locals() and isinstance(X, pd.DataFrame): # Tenta obter a estrutura original do DataFrame se disponível
         X_train = pd.DataFrame(X_train, columns=X.columns)
         X_test = pd.DataFrame(X_test, columns=X.columns)
         print("Convertido X_train e X_test para DataFrames com base nas colunas originais de X.")
    else:
         print("\nErro: Não é possível converter X_train/X_test para DataFrames sem informações de coluna originais. Pulando pré-processamento detalhado.")
         X_train_processed = X_train # Mantém como array/formato existente
         X_test_processed = X_test # Mantém como array/formato existente
         feature_names = [f'Feature {i}' for i in range(X_train.shape[1])] # Usa nomes genéricos
         preprocessing_successful = False
         print("-" * 50) # Separa mensagens de pré-processamento

# Se a conversão para DataFrame foi bem-sucedida ou X_train/X_test já eram DataFrames:
if isinstance(X_train, pd.DataFrame) and isinstance(X_test, pd.DataFrame):
    preprocessing_successful = True
    X_train_processed = X_train.copy() # Cria cópias para evitar modificar as divisões originais
    X_test_processed = X_test.copy()
    y_train_processed = y_train.copy()
    y_test_processed = y_test.copy()

    # 1. Feature Engineering (Exemplo simples: Relação Banheiro por Área)
    feature_eng_cols = ['banheiro-min', 'area(m2)-min']
    if all(col in X_train_processed.columns for col in feature_eng_cols):
        X_train_processed['banheiro_por_area'] = pd.to_numeric(X_train_processed['banheiro-min'], errors='coerce') / (pd.to_numeric(X_train_processed['area(m2)-min'], errors='coerce') + 1e-6)
        X_test_processed['banheiro_por_area'] = pd.to_numeric(X_test_processed['banheiro-min'], errors='coerce') / (pd.to_numeric(X_test_processed['area(m2)-min'], errors='coerce') + 1e-6)

        initial_train_size_fe = X_train_processed.shape[0]
        X_train_processed.dropna(subset=['banheiro_por_area'], inplace=True)
        y_train_processed = y_train_processed.loc[X_train_processed.index]
        X_test_processed.dropna(subset=['banheiro_por_area'], inplace=True)
        y_test_processed = y_test_processed.loc[X_test_processed.index]

        print("\nNova feature 'banheiro_por_area' criada e NaNs removidos.")
        print(f"Novo tamanho do conjunto de treino após FE e remoção de NaNs: {X_train_processed.shape}")
    else:
        print(f"\nColunas {feature_eng_cols} não encontradas para Feature Engineering. Pulando FE.")


    # 2. Tratamento de Outliers (Exemplo básico: Remoção baseada em Z-score em features selecionadas)
    print("\nVerificando e removendo outliers em features selecionadas no conjunto de treino:")
    if isinstance(X_train_processed, pd.DataFrame):
        if 'preco' in X_train_processed.columns:
             X_train_temp = X_train_processed.drop(columns=['preco'])
        else:
             X_train_temp = X_train_processed.copy()

        y_train_aligned = y_train_processed.loc[X_train_temp.index]
        train_combined = pd.concat([X_train_temp, y_train_aligned], axis=1)

        initial_train_size = train_combined.shape[0]
        outlier_indices_combined = pd.Index([])
        outlier_threshold = 3

        features_to_check_outliers_combined = [col for col in train_combined.columns if pd.api.types.is_numeric_dtype(train_combined[col])]

        for feature in features_to_check_outliers_combined:
            if feature in train_combined.columns:
                std_dev = train_combined[feature].std()
                if std_dev > 0:
                    z_scores_feature = np.abs((train_combined[feature] - train_combined[feature].mean()) / std_dev)
                    feature_outlier_indices = train_combined.index[z_scores_feature > outlier_threshold]
                    outlier_indices_combined = outlier_indices_combined.union(feature_outlier_indices)

        if not outlier_indices_combined.empty:
            X_train_cleaned = X_train_processed.drop(outlier_indices_combined)
            y_train_cleaned = y_train_processed.drop(outlier_indices_combined)
            print(f"\nRemovidas {initial_train_size - X_train_cleaned.shape[0]} linhas de treino contendo outliers (Z-score > {outlier_threshold}).")
            print(f"Novo tamanho do conjunto de treino após remoção de outliers: {X_train_cleaned.shape}")
            X_train_processed = X_train_cleaned
            y_train_processed = y_train_cleaned
        else:
            print("\nNenhum outlier identificado com base em Z-score. Nenhuma linha removida.")
    else:
        print("\nX_train_processed não é um pandas DataFrame. Pulando Tratamento de Outliers.")


    # 3. Seleção de Features (Exemplo básico: Selecionar K melhores features usando f_regression)
    print("\n--- Seleção de Features ---")
    if isinstance(X_train_processed, pd.DataFrame):
        X_train_numeric = X_train_processed.select_dtypes(include=np.number)

        k = 10
        if X_train_numeric.shape[1] > 0 and X_train_numeric.shape[1] >= k:
            try:
                selector = SelectKBest(score_func=f_regression, k=k)
                if np.isfinite(y_train_processed).all():
                     selector.fit(X_train_numeric, y_train_processed)
                else:
                     print("\nAviso: Valores não finitos encontrados em y_train_processed. Pulando seleção de features.")
                     selector = None

                if selector:
                    selected_features_indices = selector.get_support(indices=True)
                    selected_features_names = X_train_numeric.columns[selected_features_indices]

                    print(f"\nSelecionadas as {k} melhores features usando SelectKBest (f_regression):")
                    print(selected_features_names.tolist())

                    if isinstance(X_test_processed, pd.DataFrame):
                        X_test_numeric = X_test_processed.select_dtypes(include=np.number)
                        X_test_selected = X_test_numeric.reindex(columns=selected_features_names, fill_value=0) # Usa reindex com fill_value=0

                        print(f"Dimensões de X_train após seleção de features: {X_train_numeric[selected_features_names].shape}")
                        print(f"Dimensões de X_test após seleção de features: {X_test_selected.shape}")
                        X_train_processed = X_train_numeric[selected_features_names]
                        X_test_processed = X_test_selected
                    else:
                        print("\nX_test_processed não é um pandas DataFrame. Pulando atualização de X_test após seleção de features.")
                        X_train_processed = X_train_numeric[selected_features_names]
                elif X_train_numeric.shape[1] > 0:
                     print(f"\nSeleção de features pulada devido a valores não finitos ou outro erro. Usando todas as features numéricas disponíveis ({X_train_numeric.shape[1]}).")
                     X_train_processed = X_train_numeric
                     if isinstance(X_test_processed, pd.DataFrame):
                          X_test_processed = X_test_processed.select_dtypes(include=np.number)
                          common_cols = list(set(X_train_processed.columns) & set(X_test_processed.columns))
                          X_train_processed = X_train_processed[common_cols]
                          X_test_processed = X_test_processed[common_cols]
                     else:
                          print("\nX_test_processed não é um pandas DataFrame. Não é possível garantir colunas comuns após pular a seleção de features.")

            except Exception as e:
                print(f"\nErro durante a seleção de features: {e}")
                print("Continuando com todas as features numéricas disponíveis após o pré-processamento.")
                X_train_processed = X_train_numeric
                if isinstance(X_test_processed, pd.DataFrame):
                    X_test_processed = X_test_processed.select_dtypes(include=np.number)
                    common_cols = list(set(X_train_processed.columns) & set(X_test_processed.columns))
                    X_train_processed = X_train_processed[common_cols]
                    X_test_processed = X_test_processed[common_cols]
                else:
                    print("\nX_test_processed não é um pandas DataFrame. Não é possível garantir colunas comuns após falha na seleção de features.")

        elif X_train_numeric.shape[1] > 0:
             print(f"\nNúmero de features numéricas disponíveis ({X_train_numeric.shape[1]}) é menor que K ({k}). Pulando seleção de features e usando todas as features numéricas disponíveis.")
             X_train_processed = X_train_numeric
             if isinstance(X_test_processed, pd.DataFrame):
                  X_test_processed = X_test_processed.select_dtypes(include=np.number)
                  common_cols = list(set(X_train_processed.columns) & set(X_test_processed.columns))
                  X_train_processed = X_train_processed[common_cols]
                  X_test_processed = X_test_processed[common_cols]
             else:
                  print("\nX_test_processed não é um pandas DataFrame. Não é possível garantir colunas comuns após pular a seleção de features.")
        else:
            print("\nNão há features numéricas disponíveis em X_train_processed para seleção. Pulando seleção de features.")
            X_train_processed = X_train_numeric # Será um DataFrame vazio
            if isinstance(X_test_processed, pd.DataFrame):
                X_test_processed = X_test_processed.select_dtypes(include=np.number)
            else:
                 pass # X_test_processed permanece como está


    else:
        print("\nX_train_processed não é um pandas DataFrame. Pulando Seleção de Features.")


    # 4. Padronização (Reaplica após pré-processamento adicional)
    print("\n--- Padronização ---")
    scaler_std = StandardScaler()

    if isinstance(X_train_processed, (pd.DataFrame, np.ndarray)) and X_train_processed.shape[1] > 0:
        try:
            scaler_std.fit(X_train_processed)
            X_train_standardized = scaler_std.transform(X_train_processed)
            if isinstance(X_test_processed, (pd.DataFrame, np.ndarray)):
                 # Garante que X_test_processed tem as mesmas colunas que X_train_processed antes de transformar
                 if isinstance(X_test_processed, pd.DataFrame) and isinstance(X_train_processed, pd.DataFrame):
                      X_test_aligned = X_test_processed.reindex(columns=X_train_processed.columns, fill_value=0)
                      X_test_standardized = scaler_std.transform(X_test_aligned)
                 else: # Se um for um array, assume que estão alinhados por posição
                      if X_test_processed.shape[1] == X_train_processed.shape[1]:
                           X_test_standardized = scaler_std.transform(X_test_processed)
                      else:
                           print("\nAviso: X_test_processed e X_train_processed têm número diferente de features após o pré-processamento. Pulando padronização de X_test.")
                           X_test_standardized = None # Define como None se não puder padronizar
            else:
                 print("\nX_test_processed não é um DataFrame ou array NumPy válido. Pulando padronização de X_test.")
                 X_test_standardized = None # Define como None se não puder padronizar


            print("\nDados padronizados.")

            # Atualiza X_train e X_test para os dados padronizados
            X_train_final = X_train_standardized
            X_test_final = X_test_standardized # X_test_final pode ser None se a padronização falhou


        except Exception as e:
            print(f"\nErro durante a padronização dos dados: {e}")
            print("Continuando sem padronização devido ao erro.")
            # Se a padronização falhar, mantém X_train_final e X_test_final como estavam antes desta etapa
            X_train_final = X_train_processed # Usa dados não padronizados se a padronização falhou
            X_test_final = X_test_processed


    else:
        print("Não há features suficientes em X_train_processed para padronização.")
        X_train_final = X_train_processed # Usa dados não padronizados
        X_test_final = X_test_processed # Usa dados não padronizados


    print("-" * 50)


# --- Treina e Avalia Modelos (Replicado) ---
print("\n--- Treinamento de Modelos ---")

# 1. Treina Modelo Linear Regression
print("\nTreinando Modelo Linear Regression:")
modelo_lr = LinearRegression()

if 'X_train_final' in locals() and isinstance(X_train_final, (pd.DataFrame, np.ndarray)) and isinstance(y_train_processed, (pd.Series, np.ndarray)):
    if X_train_final.shape[0] > 0 and X_train_final.shape[1] > 0 and len(y_train_processed) == X_train_final.shape[0]:
        try:
            modelo_lr.fit(X_train_final, y_train_processed)
            print("Modelo Linear Regression treinado com sucesso.")
        except Exception as e:
            print(f"\nErro durante o treinamento da Linear Regression: {e}")
            print("Pulando treinamento da Linear Regression.")
            modelo_lr = None # Define como None se o treinamento falhar
    else:
        print("Não há dados válidos em X_train_final ou y_train_processed para treinar a Linear Regression.")
        modelo_lr = None
else:
    print("\nX_train_final ou y_train_processed não são DataFrames/arrays válidos. Pulando treinamento da Linear Regression.")
    modelo_lr = None


# 2. Treina Modelo RandomForestRegressor
print("\nTreinando Modelo RandomForestRegressor:")
modelo_rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

if 'X_train_final' in locals() and isinstance(X_train_final, (pd.DataFrame, np.ndarray)) and isinstance(y_train_processed, (pd.Series, np.ndarray)):
    if X_train_final.shape[0] > 0 and X_train_final.shape[1] > 0 and len(y_train_processed) == X_train_final.shape[0]:
        try:
            modelo_rf.fit(X_train_final, y_train_processed)
            print("Modelo RandomForestRegressor treinado com sucesso.")
        except Exception as e:
             print(f"\nErro durante o treinamento do RandomForestRegressor: {e}")
             print("Pulando treinamento do RandomForestRegressor.")
             modelo_rf = None # Define como None se o treinamento falhar
    else:
        print("Não há dados válidos em X_train_final ou y_train_processed para treinar o RandomForestRegressor.")
        modelo_rf = None
else:
    print("\nX_train_final ou y_train_processed não são DataFrames/arrays válidos. Pulando treinamento do RandomForestRegressor.")
    modelo_rf = None


print("-" * 50)


# --- Interpretação do Modelo ---
print("\n--- Interpretação do Modelo ---")

# Interpretação dos Coeficientes da Linear Regression
if modelo_lr is not None and hasattr(modelo_lr, 'coef_'):
    print("\nCoeficientes da Linear Regression (em dados padronizados, se a padronização foi bem-sucedida):")
    # Tenta obter nomes das features a partir do DataFrame X_train_processed antes da padronização
    if 'X_train_processed' in locals() and isinstance(X_train_processed, pd.DataFrame):
        feature_names = X_train_processed.columns
        coefficients = modelo_lr.coef_

        # Garante que o número de nomes de features corresponde ao número de coeficientes
        if len(feature_names) == len(coefficients):
            for feature, coef in zip(feature_names, coefficients):
                print(f"{feature}: {coef:.4f}")
        else:
             print(f"Aviso: O número de nomes de features ({len(feature_names)}) não corresponde ao número de coeficientes ({len(coefficients)}). Não é possível exibir os nomes corretamente.")
             for i, coef in enumerate(coefficients):
                  print(f"Feature {i}: {coef:.4f}")

        print(f"Intercepto: {modelo_lr.intercept_:.4f}")

        # Visualizando coeficientes (opcional)
        if len(feature_names) == len(coefficients): # Apenas plota se os nomes corresponderem aos coeficientes
             plt.figure(figsize=(12, 6))
             coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
             coef_df['Abs_Coefficient'] = np.abs(coef_df['Coefficient'])
             coef_df = coef_df.sort_values(by='Abs_Coefficient', ascending=False)

             sns.barplot(x='Coefficient', y='Feature', data=coef_df)
             plt.title('Coeficientes da Linear Regression (em dados padronizados)')
             plt.xlabel('Valor do Coeficiente')
             plt.ylabel('Feature')
             plt.show()
        else:
             print("\nPulando visualização dos coeficientes da Linear Regression devido à incompatibilidade entre nomes de features e coeficientes.")

    else:
        print("\nNão foi possível recuperar nomes de features para os coeficientes da Linear Regression.")
        # Se os nomes das features não puderem ser recuperados, apenas imprime os coeficientes com índice
        if hasattr(modelo_lr, 'coef_'):
            coefficients = modelo_lr.coef_
            for i, coef in enumerate(coefficients):
                print(f"Feature {i}: {coef:.4f}")
            print(f"Intercepto: {modelo_lr.intercept_:.4f}")


else:
     print("\nModelo Linear Regression não treinado ou coeficientes não disponíveis para interpretação.")


# Importância das Features para o RandomForestRegressor
if modelo_rf is not None and hasattr(modelo_rf, 'feature_importances_'):
    print("\nImportância das Features (RandomForestRegressor):")
    feature_importances = modelo_rf.feature_importances_

    # Tenta obter nomes das features a partir do DataFrame X_train_processed antes da padronização
    if 'X_train_processed' in locals() and isinstance(X_train_processed, pd.DataFrame):
         feature_names = X_train_processed.columns

         if len(feature_names) == len(feature_importances): # Garante que os nomes correspondem às importâncias
              importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
              importance_df = importance_df.sort_values(by='Importance', ascending=False)

              for index, row in importance_df.iterrows():
                  print(f"{row['Feature']}: {row['Importance']:.4f}")

              plt.figure(figsize=(12, 7))
              sns.barplot(x='Importance', y='Feature', data=importance_df)
              plt.title('Importância das Features (RandomForestRegressor)')
              plt.xlabel('Importância')
              plt.ylabel('Feature')
              plt.show()
         else:
              print("\nNão foi possível obter a Importância das Features para o RandomForestRegressor: Nomes das features não correspondem às importâncias.")
              # Se os nomes não corresponderem, apenas imprime as importâncias com índice
              for i, importance in enumerate(feature_importances):
                   print(f"Feature {i}: {importance:.4f}")
    else:
        print("\nNão foi possível recuperar nomes de features para a importância das features do RandomForestRegressor.")
        # Se os nomes das features não puderem ser recuperados, apenas imprime as importâncias com índice
        for i, importance in enumerate(feature_importances):
            print(f"Feature {i}: {importance:.4f}")

else:
    print("\nModelo RandomForestRegressor não treinado ou importância das features não disponível para interpretação.")


print("-" * 50)
```

Nossa Conclusão: A interpretação dos coeficientes da Regressão Linear (em dados padronizados) nos mostra a magnitude e direção do impacto de cada feature no preço. Por exemplo, um coeficiente positivo alto para 'area(m2)-max' indica que um aumento na área (em unidades de desvio padrão) está associado a um aumento significativo no preço. A importância das features do RandomForestRegressor, por outro lado, indica quais variáveis foram mais utilizadas pelo modelo para fazer suas previsões. Ambos os métodos nos ajudam a identificar as "pistas" mais relevantes para prever o valor de venda dos imóveis, destacando o IPTU, área e número de banheiros como fatores chave neste dataset.
