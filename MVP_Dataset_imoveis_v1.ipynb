```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # Para o modelo de regressão
from sklearn.metrics import mean_squared_error, r2_score # Para avaliação do modelo

# Opcionais para pré-processamento mais avançado, não usados neste MVP, mas mantidos do template
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


# URL da base de dados no GitHub
github_csv_url = 'https://raw.githubusercontent.com/felipee51/FelipeCavalcante/refs/heads/main/Base_Imoveis_RJ_tratada.csv'

try:
    # Tenta carregar o DataFrame a partir do URL do GitHub, especificando o delimitador
    df = pd.read_csv(github_csv_url, delimiter=';')
    print(f"Dados carregados com sucesso do GitHub: {github_csv_url}")
    print("Primeiras 5 linhas dos dados originais:")
    print(df.head())
    print(f"Total de {len(df)} imóveis carregados.")

    # Renomeia a coluna 'valor' para 'preco', que será a nossa variável alvo.
    target_column_original_name = 'valor'
    target_column_model_name = 'preco'

    if target_column_original_name not in df.columns:
        raise ValueError(f"A coluna de preço esperada '{target_column_original_name}' não foi encontrada no CSV.")

    df = df.rename(columns={target_column_original_name: target_column_model_name})

    # Explicitly drop non-numeric columns that are not used in this MVP
    columns_to_drop = ['local', 'bairro/cidade']
    df = df.drop(columns=columns_to_drop, errors='ignore') # 'errors='ignore' prevents error if columns are not found


    # Converte todas as colunas que podem ser numéricas para o tipo numérico.
    # Valores não numéricos serao convertidos para NaN.
    for col in df.columns:
        # Trata colunas com intervalos como 'area(m2)-min' e 'area(m2)-max'
        if '-min' in col or '-max' in col:
            df[col] = pd.to_numeric(df[col].astype(str).str.replace('[^0-9.]', '', regex=True), errors='coerce')
        else:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Check for nulls before dropping
    print("\nValores nulos no dataset antes de remover linhas:")
    print(df.isnull().sum())

    # Remove linhas que contenham valores ausentes (NaN) nas colunas apos a conversao.
    # Inserido para a regressão linear, que nao lida com NaNs.
    df.dropna(inplace=True)


    print(f"\nColuna '{target_column_original_name}' renomeada para '{target_column_model_name}'.")
    print("Todas as colunas foram convertidas para numérico, e linhas com valores ausentes foram removidas.")
    print(f"Total de {len(df)} imóveis após a limpeza e preparação.")
    print("Primeiras 5 linhas dos dados preparados:")
    print(df.head())

except Exception as e:
    print(f"Erro ao carregar ou processar os dados do GitHub: {e}")
    print("Caindo para a geração de dados sintéticos para demonstração (fallback).")
    # --- Geração de Dados Sintéticos (Fallback se o GitHub falhar ou para teste) ---
    np.random.seed(42) # Para reprodutibilidade
    num_imoveis = 100
    # Gerando features comuns que se assemelham ao seu dataset do RJ
    area_sintetica = np.random.randint(40, 500, num_imoveis)
    quartos_sinteticos = np.random.randint(1, 6, num_imoveis)
    suites_sinteticas = np.random.randint(0, 4, num_imoveis)
    vagas_sinteticas = np.random.randint(0, 3, num_imoveis)
    banheiros_sinteticos = np.random.randint(1, 5, num_imoveis)
    distancia_centro_sintetica = np.random.uniform(1, 20, num_imoveis)
    valor_condominio_sintetico = np.random.randint(200, 2000, num_imoveis)
    valor_iptu_sintetico = np.random.randint(50, 500, num_imoveis)

    # Preço sintético baseado em algumas features
    precos_sinteticos = (
        1000 * area_sintetica +
        50000 * quartos_sinteticos +
        30000 * suites_sinteticas +
        20000 * vagas_sinteticas +
        40000 * banheiros_sinteticos -
        5000 * distancia_centro_sintetica +
        0.5 * valor_condominio_sintetico + # Condominio pode influenciar um pouco
        0.1 * valor_iptu_sintetico +      # IPTU pode influenciar um pouco
        np.random.normal(0, 100000, num_imoveis) # Ruído aleatório
    )
    # Garantir preços positivos
    precos_sinteticos[precos_sinteticos < 1000] = 1000 + np.random.normal(0, 500, sum(precos_sinteticos < 1000))

    data_sintetica = {
        'area': area_sintetica,
        'quartos': quartos_sinteticos,
        'suites': suites_sinteticas,
        'vagas': vagas_sinteticas,
        'banheiros': banheiros_sinteticos,
        'distancia_centro': distancia_centro_sintetica,
        'valor_condominio': valor_condominio_sintetico,
        'valor_iptu': valor_iptu_sintetico,
        'preco': precos_sinteticos
    }
    df = pd.DataFrame(data_sintetica)
    print("Dados sintéticos gerados como fallback.")
    print(df.head())

# Verifica se o DataFrame foi carregado/creado com sucesso e possui a coluna 'preco'
if 'df' not in locals() or df.empty or target_column_model_name not in df.columns:
    print("\nERRO CRÍTICO: Não foi possível obter um DataFrame válido com a coluna de preço.")
    print("Por favor, verifique o URL do GitHub, o conteúdo do CSV ou o bloco de geração de dados sintéticos.")
    exit() # Interrompe a execução se não houver dados válidos

print("-" * 50)

# Análise de Dados

# Nesta etapa de Análise de Dados Exploratória (EDA) sobre o dataset de imóveis, visamos entender a distribuição, as relações e as características das variáveis, o que é crucial para as etapas subsequentes de pré-processamento e modelagem.
# Total e Tipo das Instâncias

print(f"Total de instâncias: {len(df)}")
print("\nTipos de dados por coluna:")
print(df.info())

# Exemplo de contagem para uma coluna categórica (se houvesse, como 'bairro' ou 'tipo_imovel' antes do dropna)
# if 'bairro' in df.columns:
#     plt.figure(figsize=(10, 7))
#     sns.countplot(y='bairro', data=df, order=df['bairro'].value_counts().index[:10]) # Top 10 bairros
#     plt.title('Distribuição dos Imóveis por Bairro (Top 10)')
#     plt.xlabel('Contagem')
#     plt.ylabel('Bairro')
#     plt.show()

# Estatísticas Descritivas

# Estatísticas descritivas fornecem um resumo das características numéricas, incluindo média, desvio padrão, mínimo, máximo e quartis.

# estatísticas descritivas básicas do dataset
print("\nEstatísticas Descritivas do Dataset:")
print(df.describe())

# Média

# A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

# média dos atributos numéricos do dataset
print("\nMédia dos atributos numéricos:")
print(df.describe().loc['mean'])

# Gráfico de barras para a média de atributos selecionados
plt.figure(figsize=(12, 6))
df.drop(columns=[target_column_model_name]).mean().plot(kind='bar')
plt.title('Média dos Atributos Numéricos')
plt.xlabel('Atributo')
plt.ylabel('Valor Médio')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Desvio Padrão

# O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

# desvio padrão dos atributos numéricos do dataset
print("\nDesvio Padrão dos atributos numéricos:")
print(df.describe().loc['std'])

# Gráfico de barras para o desvio padrão de atributos selecionados
plt.figure(figsize=(12, 6))
df.drop(columns=[target_column_model_name]).std().plot(kind='bar', color='orange')
plt.title('Desvio Padrão dos Atributos Numéricos')
plt.xlabel('Atributo')
plt.ylabel('Desvio Padrão')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Histograma

# A distribuição de dados descreve como os valores de uma variável se espalham, ou seja, a frequência com que diferentes valores ocorrem. Entender a distribuição é crucial na análise de dados, pois revela padrões, tendências centrais, dispersão e a presença de valores atípicos (outliers). O histograma é uma ferramenta visual fundamental para representar essa distribuição, mostrando a forma dos dados, se são simétricos ou assimétricos, unimodais ou multimodais.
# Distribuição do area

# Use a coluna 'area(m2)-min' para o histograma de área se ela existir e for numérica
area_column_to_plot = 'area(m2)-min' if 'area(m2)-min' in df.columns and pd.api.types.is_numeric_dtype(df['area(m2)-min']) else None

if area_column_to_plot and not df[area_column_to_plot].isnull().all(): # Check if column exists, is numeric, and not all values are NaN
    plt.figure(figsize=(8, 6))
    sns.histplot(df[area_column_to_plot].dropna(), kde=True) # Drop NaNs for plotting
    plt.title(f'Distribuição da Área dos Imóveis ({area_column_to_plot})')
    plt.xlabel('Área (m²)')
    plt.ylabel('Frequência')
    plt.show()
else:
    print(f"\nColuna '{area_column_to_plot}' não encontrada, não numérica ou contém apenas valores ausentes para visualização.")


# Distribuição do quartos

# Use a coluna 'quarto-min' para o histograma de quartos se ela existir e for numérica
quartos_column_to_plot = 'quarto-min' if 'quarto-min' in df.columns and pd.api.types.is_numeric_dtype(df['quarto-min']) else None

if quartos_column_to_plot and not df[quartos_column_to_plot].isnull().all(): # Check if column exists, is numeric, and not all values are NaN
    plt.figure(figsize=(8, 6))
    quartos_data = df[quartos_column_to_plot].dropna() # Drop NaNs for plotting
    if not quartos_data.empty:
        bins_range = range(int(quartos_data.min()), int(quartos_data.max()) + 2)
        sns.histplot(quartos_data, kde=False, bins=bins_range)
        plt.title(f'Distribuição do Número de Quartos ({quartos_column_to_plot})')
        plt.xlabel('Número de Quartos')
        plt.ylabel('Frequência')
        plt.xticks(range(int(quartos_data.min()), int(quartos_data.max()) + 1))
        plt.show()
    else:
         print(f"\nColuna '{quartos_column_to_plot}' contém apenas valores ausentes após a limpeza para visualização.")
else:
     print(f"\nColuna '{quartos_column_to_plot}' não encontrada, não numérica ou contém apenas valores ausentes para visualização.")


# Boxplot

# Boxplots são excelentes para visualizar a distribuição de uma variável numérica em relação a uma categórica (se tivéssemos tratado 'bairro' ou 'tipo_imovel' adequadamente), ou para identificar outliers. Para este MVP, focaremos em scatter plots para relações entre features numéricas e o preço.
# Relação area vs. preco

# Use a coluna 'area(m2)-min' para o scatter plot area vs preco se ela existir e for numérica
area_column_for_scatter = 'area(m2)-min' if 'area(m2)-min' in df.columns and pd.api.types.is_numeric_dtype(df['area(m2)-min']) else None

if area_column_for_scatter and target_column_model_name in df.columns and not df[[area_column_for_scatter, target_column_model_name]].isnull().all().any(): # Check if columns exist, are numeric, and not all values are NaN
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=area_column_for_scatter, y=target_column_model_name, data=df.dropna(subset=[area_column_for_scatter, target_column_model_name]), alpha=0.6) # Drop NaNs for plotting
    plt.title(f'Valor de Aluguel vs. Área do Imóvel ({area_column_for_scatter})')
    plt.xlabel('Área (m²)')
    plt.ylabel('Valor de Aluguel (R$)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
else:
    print(f"\nColunas '{area_column_for_scatter}' ou '{target_column_model_name}' não encontradas, não numéricas ou contêm apenas valores ausentes para visualização de scatter plot.")


# Relação quartos vs. preco

# Use a coluna 'quarto-min' para o boxplot quartos vs preco if it exists and is numeric
quartos_column_for_boxplot = 'quarto-min' if 'quarto-min' in df.columns and pd.api.types.is_numeric_dtype(df['quarto-min']) else None


if quartos_column_for_boxplot and target_column_model_name in df.columns and not df[[quartos_column_for_boxplot, target_column_model_name]].isnull().all().any(): # Check if columns exist, are numeric, and not all values are NaN
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=quartos_column_for_boxplot, y=target_column_model_name, data=df.dropna(subset=[quartos_column_for_boxplot, target_column_model_name])) # Drop NaNs for plotting
    plt.title(f'Valor de Aluguel por Número de Quartos ({quartos_column_for_boxplot})')
    plt.xlabel('Número de Quartos')
    plt.ylabel('Valor de Aluguel (R$)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
else:
    print(f"\nColunas '{quartos_column_for_boxplot}' ou '{target_column_model_name}' não encontradas, não numéricas ou contêm apenas valores ausentes para visualização de boxplot.")

# Matriz de Correlação

# A matriz de correlação mede a força e a direção de uma relação linear entre os atributos numéricos. Valores próximos a 1 indicam uma forte correlação positiva, -1 uma forte correlação negativa, e 0 ausência de correlação linear.

# Matriz de correlação para todas as colunas numéricas
# Garante que apenas colunas numéricas sejam incluídas
df_numeric = df.select_dtypes(include=np.number)
print("\nMatriz de Correlação das Características Numéricas:")
print(df_numeric.corr())

plt.figure(figsize=(10, 8))
# mapa de calor das variáveis numéricas
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação das Características Numéricas do Dataset de Imóveis')
plt.show()

# O mapa de calor da matriz de correlação revela as relações entre as variáveis numéricas, incluindo a correlação das features com o preco. Isso é crucial para entender quais características têm maior impacto no valor do aluguel.
# Tratamento de Valores Nulos

# O tratamento de valores nulos é crucial e pode envolver imputação (preenchimento com média, mediana, moda) ou remoção de linhas/colunas. Neste código, já realizamos df.dropna(inplace=True) após a conversão de tipo, garantindo que não há valores nulos no DataFrame final df antes da modelagem.

# Verificar a presença de valores nulos no dataset após a limpeza
print("\nValores nulos no dataset de Imóveis após a limpeza:")
print(df.isnull().sum())

# Pré-Processamento de Dados

# O pré-processamento de dados é uma etapa crucial para preparar os dados para modelagem, garantindo que estejam no formato correto e otimizados para o desempenho do algoritmo.

# Separar features (X) e target (y)
target_column_model_name = 'preco' # Definido novamente para clareza na seção
y = df[target_column_model_name]

# As features (X) serão TODAS as outras colunas numéricas no DataFrame, exceto a coluna de preço.
X = df.drop(columns=[target_column_model_name])
# Garante que X contém apenas colunas numéricas
X = X.select_dtypes(include=np.number)

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Usando 30% para teste

print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")

# Normalização

# A normalização escala os dados para um intervalo fixo, geralmente entre 0 e 1. É útil quando o algoritmo de machine learning assume que as características estão em uma escala semelhante.

# Inicializar o MinMaxScaler
scaler_norm = MinMaxScaler()

# Aprende min e max APENAS de X_train
scaler_norm.fit(X_train)
X_train_normalized = scaler_norm.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_normalized = scaler_norm.transform(X_test)

# Exibir as primeiras linhas dos dados normalizados (como DataFrame para melhor visualização)
df_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)

print("\nPrimeiras 5 linhas dos dados normalizados (treino):")
print(df_normalized.head())

# Visualização da distribuição após a normalização (exemplo para uma característica)
if 'area(m2)-min' in df_normalized.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_normalized['area(m2)-min'], kde=True)
    plt.title('Distribuição da Área (Normalizado)')
    plt.xlabel('Área Normalizada')
    plt.ylabel('Frequência')
    plt.show()
else:
    print("\nColuna 'area(m2)-min' não encontrada para visualização após normalização.")

# O histograma da área após a normalização mostra que os valores foram escalados para o intervalo de 0 a 1, mantendo a forma da distribuição original.
# Padronização

# A padronização (ou Z-score scaling) transforma os dados para ter média 0 e desvio padrão 1. É útil para algoritmos que são sensíveis à escala das características, como SVMs ou redes neurais.

# Inicializar o StandardScaler
scaler_std = StandardScaler()

# Aprende média e desvio padrão APENAS de X_train
scaler_std.fit(X_train)
X_train_standardized = scaler_std.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_standardized = scaler_std.transform(X_test)

# Exibir as primeiras linhas dos dados padronizados (como DataFrame para melhor visualização)
df_standardized = pd.DataFrame(X_train_standardized, columns=X_train.columns)

print("\nPrimeiras 5 linhas dos dados padronizados (treino):")
print(df_standardized.head())

# Visualização da distribuição após a padronização (exemplo para uma característica)
if 'area(m2)-min' in df_standardized.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_standardized['area(m2)-min'], kde=True)
    plt.title('Distribuição da Área (Padronizado)')
    plt.xlabel('Área Padronizada')
    plt.ylabel('Frequência')
    plt.show()
else:
    print("\nColuna 'area(m2)-min' não encontrada para visualização após padronização.")

# O histograma da área após a padronização mostra que os valores foram transformados para ter uma média próxima de zero e um desvio padrão de um, centralizando a distribuição.
# Outras Transformações e Etapas de Pré-Processamento

# Para este problema de previsão de aluguel, outras etapas cruciais seriam:

#     One-Hot Encoding para variáveis categóricas: As colunas bairro e tipo_imovel são categóricas e possuem informações valiosas. Elas precisariam ser convertidas para um formato numérico que o modelo possa usar (e.g., One-Hot Encoding ou Label Encoding).

#     Feature Engineering: Criar novas features a partir das existentes, como relacao_suite_quarto, densidade_banheiros (banheiros por área), ou idade_imovel se houvesse uma data de construção/publicação.

#     Remoção de Outliers: Identificar e tratar valores extremos no valor_aluguel ou em outras features, que podem distorcer o treinamento do modelo.

#     Seleção de Features: Utilizar técnicas para identificar as features mais relevantes para a previsão, potencialmente removendo as menos informativas para melhorar a performance e interpretabilidade do modelo.

# Modelagem e Avaliação

# Nesta seção, treinamos um modelo de Regressão Linear com os dados padronizados (uma boa prática para modelos baseados em distância/gradiente) e avaliamos seu desempenho.

# --- 3. Treinamento do Modelo ---
# Usaremos um modelo de Regressão Linear.

print("\n--- 3. Treinamento do Modelo ---")
modelo = LinearRegression()
modelo.fit(X_train_standardized, y_train) # Usando dados padronizados para treinamento

print("Modelo de Regressão Linear treinado com sucesso.")
# Para interpretar os coeficientes, lembre-se que eles são baseados nos dados padronizados.
# Coeficientes: o impacto de uma unidade de desvio padrão da feature no preço.
# print(f"Coeficientes do modelo para cada feature: {modelo.coef_}") # Comentado para evitar output muito longo
print(f"Intercepto do modelo: {modelo.intercept_}")
print("-" * 50)

# --- 4. Avaliação do Modelo ---
# Fazer previsões no conjunto de teste (também padronizado) e calcular métricas de erro.

print("\n--- 4. Avaliação do Modelo ---")
y_pred_test = modelo.predict(X_test_standardized) # Previsões no conjunto de teste padronizado

# Erro Quadrático Médio (MSE): Média dos quadrados dos erros. Quanto menor, melhor.
mse = mean_squared_error(y_test, y_pred_test)
# R² (Coeficiente de Determinacao): Proporcao da variancia na variavel dependente que é previsivel a partir das variaveis independentes. Quanto mais proximo de 1, melhor.
r2 = r2_score(y_test, y_pred_test)

print(f"Erro Quadrático Médio (MSE) no conjunto de TESTE: {mse:.2f}")
print(f"R² (Coeficiente de Determinação) no conjunto de TESTE: {r2:.2f}")
print("-" * 50)

# --- 5. Realizar Previsoes (Exemplo) ---
# Usar o modelo treinado para prever o preco de um novo imovel.
# Para o exemplo, usaremos a media das features do conjunto de treinamento padronizado.

print("\n--- 5. Realizar Previsões (Exemplo) ---")
# Cria um novo DataFrame com a média das features de treinamento (já padronizadas)
# Precisa despadronizar os valores médios para mostrar as características no formato original
# mas o modelo espera a entrada padronizada.
# A melhor prática para o exemplo é pegar uma linha real do X_test e prever.
# Ou, criar um "novo imóvel" com valores "reais" e depois padronizá-los.

# Vamos criar um "novo imóvel" com valores em escala original para o exemplo
# e depois padronizá-lo para a previsão.
# Ajustando colunas para o exemplo de imóvel com base nas colunas do dataset real
example_columns = [col for col in X_train.columns if col not in ['local', 'bairro/cidade']] # Exclui colunas não numéricas/tratadas ainda
exemplo_imovel_original = pd.DataFrame(columns=example_columns)

# Preencher com valores de exemplo (ajuste conforme as colunas numéricas do seu X_train)
# Este é um exemplo simplificado. Em um app real, o usuário preencheria esses valores.
# Use a média de X_train para colunas numéricas que não estão explicitamente definidas no exemplo
exemplo_imovel_data = X_train.mean().to_dict()

# Sobrescrever com valores específicos para o exemplo, garantindo que as chaves existem
if 'area(m2)-min' in exemplo_imovel_data: exemplo_imovel_data['area(m2)-min'] = 100
if 'quarto-min' in exemplo_imovel_data: exemplo_imovel_data['quarto-min'] = 2
if 'suite' in exemplo_imovel_data: exemplo_imovel_data['suite'] = 1
if 'garagem-min' in exemplo_imovel_data: exemplo_imovel_data['garagem-min'] = 1
if 'banheiro-min' in exemplo_imovel_data: exemplo_imovel_data['banheiro-min'] = 2
# Adicione outras colunas numéricas conforme necessário, verificando sua existência
if 'iptu' in exemplo_imovel_data: exemplo_imovel_data['iptu'] = 100
if 'condominio' in exemplo_imovel_data: exemplo_imovel_data['condominio'] = 500


exemplo_imovel_original = pd.DataFrame([exemplo_imovel_data])


# Padroniza o exemplo de imóvel usando o scaler trainado no X_train
exemplo_imovel_standardized = scaler_std.transform(exemplo_imovel_original)
exemplo_imovel_df_standardized = pd.DataFrame(exemplo_imovel_standardized, columns=X_train.columns)


preco_previsto_exemplo = modelo.predict(exemplo_imovel_df_standardized)

print(f"Características de um imóvel de exemplo (em escala original):")
print(exemplo_imovel_original.iloc[0].to_dict())
print(f"Preço de aluguel previsto para este imóvel de exemplo: R$ {preco_previsto_exemplo[0]:.2f}")
print("-" * 50)

# --- 6. Visualização de Resultados (Todas as linhas válidas) ---
# Fazer previsões em TODO o DataFrame X e comparar com TODO o DataFrame y.

print("\n--- 6. Visualização de Resultados (Com todas as linhas válidas) ---")

# Prever preços para TODAS as features (X) após padronização
# Para fazer previsões em todo X, X também precisa ser padronizado.
X_standardized_total = scaler_std.transform(X)
y_pred_total = modelo.predict(X_standardized_total)

plt.figure(figsize=(10, 7))
# O gráfico agora usa 'y' (todos os valores reais) e 'y_pred_total' (todas as previsões)
sns.scatterplot(x=y, y=y_pred_total, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2) # Linha de 45 graus ideal
plt.title('Preços Reais vs. Preços Previstos (Todas as Linhas Válidas)')
plt.xlabel('Preço Real (R$)')
plt.ylabel('Preço Previsto (R$)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

print("\nMVP de Previsão de Preços de Imóveis concluído.")
print("O gráfico agora mostra a relação entre os preços reais e os preços previstos pelo modelo para TODAS as linhas válidas do seu conjunto de dados.")
```

## 3. Pré-Processamento de Dados: Preparando o Terreno para o Modelo


O pré-processamento de dados é uma etapa crucial para preparar os dados para modelagem, garantindo que estejam no formato correto e otimizados para o desempenho do algoritmo.  

Antes de apresentarmos as pistas ao nosso "agente preditivo" (o modelo de machine learning), precisamos prepará-las. O pré-processamento é como refinar as evidências, garantindo que estejam no formato ideal para que o algoritmo possa "entender" e aprender de forma mais eficaz. É a ponte entre os dados brutos e a inteligência do modelo, o momento em que o detetive organiza as informações para a apresentação final.

### Separação de Features e Target

Separamos as variáveis independentes (features, X) da variável dependente (target, y), que é o preço do imóvel.

```python
# Separar features (X) e target (y)
target_column_model_name = 'preco' # Definido novamente para clareza na seção
y = df[target_column_model_name]

# As features (X) serão TODAS as outras colunas numéricas no DataFrame, exceto a coluna de preço.
X = df.drop(columns=[target_column_model_name])
# Garante que X contém apenas colunas numéricas
X = X.select_dtypes(include=np.number)

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Usando 30% para teste

print(f"Dimensões de X_train: {X_train.shape}")
print(f"Dimensões de X_test: {X_test.shape}")
print(f"Dimensões de y_train: {y_train.shape}")
print(f"Dimensões de y_test: {y_test.shape}")
```

Nossa Conclusão: Dividimos nossas evidências em dois grupos: um para treinamento (onde o agente aprenderá) e outro para teste (onde avaliamos o que ele aprendeu com novas evidências). Isso garante que o modelo seja avaliado de forma justa, com dados que ele nunca viu antes, simulando um cenário real de predição. A proporção 70/30 para treino/teste é um ponto de partida comum, proporcionando um bom equilíbrio entre volume de dados para aprendizado e para validação

### Normalização: Ajustando as Escalas das Pistas

A normalização escala os dados para um intervalo fixo, geralmente entre 0 e 1. É útil quando o algoritmo de machine learning assume que as características estão em uma escala semelhante.

A normalização é como nivelar o campo de jogo para nossas pistas numéricas. Ela escala todos os valores para um intervalo fixo (geralmente entre 0 e 1). Isso é crucial para algoritmos que são sensíveis à magnitude dos números (como redes neurais), garantindo que uma feature com valores grandes (como área) não domine uma feature com valores pequenos (como número de quartos) injustamente. É como garantir que todas as evidências estejam na mesma "linguagem" para o modelo.

```python
# Inicializar o MinMaxScaler
scaler_norm = MinMaxScaler()

# Aprende min e max APENAS de X_train
scaler_norm.fit(X_train)
X_train_normalized = scaler_norm.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_normalized = scaler_norm.transform(X_test)

# Exibir as primeiras linhas dos dados normalizados (como DataFrame para melhor visualização)
df_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)

print("\nPrimeiras 5 linhas dos dados normalizados (treino):")
display(df_normalized.head())

# Visualização da distribuição após a normalização (exemplo para uma característica)
if 'area(m2)-min' in df_normalized.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_normalized['area(m2)-min'], kde=True)
    plt.title('Distribuição da Área (Normalizado)')
    plt.xlabel('Área Normalizada')
    plt.ylabel('Frequência')
    plt.show()
else:
    print("\nColuna 'area(m2)-min' não encontrada para visualização após normalização.")

print("\nO histograma da área após a normalização mostra que os valores foram escalados para o intervalo de 0 a 1, mantendo a forma da distribuição original.")
```


Nossa Conclusão: O histograma da área após a normalização mostra que os valores foram escalados para o intervalo de 0 a 1, mantendo a forma da distribuição original. Isso significa que, embora os números tenham mudado, a informação que eles contêm (sua distribuição) permanece intacta, o que é vital para algoritmos que se baseiam em distâncias, assegurando que o modelo não seja "enganado" por escalas diferentes.

### Padronização: Centralizando e Padronizando a Variação


A padronização (ou Z-score scaling) transforma os dados para ter média 0 e desvio padrão 1.

Esta técnica de nivelamento transforma os dados para que tenham uma média de 0 e um desvio padrão de 1. Isso é particularmente útil para algoritmos que se beneficiam de dados centrados no zero, como a Regressão Linear, pois ajuda a convergência e o desempenho do algoritmo ao garantir que todas as features contribuam igualmente. É como ajustar a "lente" para ver todas as evidências com a mesma clareza.

```python
# Inicializar o StandardScaler
scaler_std = StandardScaler()

# Aprende média e desvio padrão APENAS de X_train
scaler_std.fit(X_train)
X_train_standardized = scaler_std.transform(X_train)
# Usa a média e o desvio padrão aprendidos de X_train
X_test_standardized = scaler_std.transform(X_test)

# Exibir as primeiras linhas dos dados padronizados (como DataFrame para melhor visualização)
df_standardized = pd.DataFrame(X_train_standardized, columns=X_train.columns)

print("\nPrimeiras 5 linhas dos dados padronizados (treino):")
display(df_standardized.head())

# Visualização da distribuição após a padronização (exemplo para uma característica)
if 'area(m2)-min' in df_standardized.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_standardized['area(m2)-min'], kde=True)
    plt.title('Distribuição da Área (Padronizado)')
    plt.xlabel('Área Padronizada')
    plt.ylabel('Frequência')
    plt.show()
else:
    print("\nColuna 'area(m2)-min' não encontrada para visualização após padronização.")

print("\nO histograma da área após a padronização mostra que os valores foram transformados para ter uma média próxima de zero e um desvio padrão de um, centralizando a distribuição.")
```

Nossa Conclusão: O histograma da área após a padronização confirma que os valores foram transformados para ter uma média próxima de zero e um desvio padrão de um, centralizando a distribuição e tornando-la mais adequada para certos modelos de aprendizado de máquina, especialmente aqueles que dependem de gradientes ou suposições de normalidade. Isso uniformiza a contribuição de cada pista para o modelo.


### Outras Transformações e Etapas de Pré-Processamento: Refinando Nossas Técnicas de Investigação


Para este problema de previsão de preço, outras etapas cruciais seriam:

Este MVP é um começo promissor, mas uma investigação completa sempre tem mais camadas.   
Para um projeto mais robusto de previsão de venda de imóveis, os próximos passos de pré-processamento seriam aprofundar as técnicas de exemplo abaixo, com um verdadeiro detetive de dados.   
Contudo, estas técnicas não foram aplicadas neste MVP.

*  **One-Hot Encoding para variáveis categóricas:** As colunas como bairro e tipo_imovel são chaves para entender o mercado do Rio. Elas contêm informações qualitativas valiosas que, para serem usadas por modelos numéricos, precisam ser convertidas. O One-Hot Encoding nos permite criar novas "pistas binárias" (0 ou 1) para cada categoria, como "É do bairro X?" ou "É um apartamento?", adicionando uma dimensão crítica de análise e melhorando significativamente a precisão do modelo.

*  **Feature Engineering (Engenharia de Características):** É a arte de criar novas "pistas" a partir das evidências existentes, como um detetive que combina fragmentos para formar uma imagem maior. Por exemplo, poderíamos calcular a relacao_suite_quarto para entender a proporção de luxo, a densidade_banheiros (banheiros por área) para avaliar a funcionalidade, ou até inferir uma idade_imovel se houvesse uma data de construção/publicação. Novas features podem revelar padrões que não são óbvios nas características originais, dando ao modelo "superpoderes" de percepção.

*  **Remoção de Outliers:** Identificar e tratar valores extremos no valor ou em outras features. Um imóvel com valor excepcionalmente alto ou baixo pode ser um "falso testemunho" que distorce o aprendizado do modelo. Lidar com eles de forma inteligente – seja removendo-os ou transformando-os – é fundamental para a robustez do modelo e para evitar conclusões precipitadas.

*  **Seleção de Features:* Utilizar técnicas para identificar as features mais relevantes para a previsão. Nem todas as pistas são igualmente importantes, e algumas podem até ser redundantes ou irrelevantes. Remover as menos informativas (ou aquelas que causam "ruído") pode melhorar a performance do modelo, reduzir o custo computacional e tornar o modelo mais interpretável, focando no que realmente importa para desvendar o caso.

## 4. Modelagem e Avaliação: Construindo e Testando Nosso Agente Preditivo


Nesta seção, treinamos um modelo de Regressão Linear com os dados padronizados (uma boa prática para modelos baseados em distância/gradiente) e avaliamos seu desempenho.

Com as pistas devidamente preparadas, é hora de construir e treinar nosso "agente preditivo" – o modelo de Regressão Linear. Em seguida, o avaliaremos para ver o quão bem ele se saiu na sua missão de previsão. ### Treinamento do Modelo

Utilizamos um modelo de Regressão Linear para prever o preço dos imóveis.

```python
print("\n--- 3. Treinamento do Modelo ---")
modelo = LinearRegression()
# Use X_train_standardized (DataFrame with feature names) for fitting
modelo.fit(X_train_standardized, y_train)

print("Modelo de Regressão Linear treinado com sucesso.")
# Para interpretar os coeficientes, lembre-se que eles são baseados nos dados padronizados.
# Coeficientes: o impacto de uma unidade de desvio padrão da feature no preço.
# print(f"Coeficientes do modelo para cada feature: {modelo.coef_}") # Comentado para evitar output muito longo
print(f"Intercepto do modelo: {modelo.intercept_}")
print("-" * 50)
```

### Avaliação do Modelo

Avaliamos o desempenho do modelo no conjunto de teste utilizando métricas comuns para regressão.

```python
print("\n--- 4. Avaliação do Modelo ---")
y_pred_test = modelo.predict(X_test_standardized) # Previsões no conjunto de teste padronizado

# Erro Quadrático Médio (MSE): Média dos quadrados dos erros. Quanto menor, melhor.
mse = mean_squared_error(y_test, y_pred_test)
# R² (Coeficiente de Determinacao): Proporcao da variancia na variavel dependente que é previsivel a partir das variaveis independentes. Quanto mais proximo de 1, melhor.
r2 = r2_score(y_test, y_pred_test)

print(f"Erro Quadrático Médio (MSE) no conjunto de TESTE: {mse:.2f}")
print(f"R² (Coeficiente de Determinação) no conjunto de TESTE: {r2:.2f}")
print("-" * 50)
```
